### Perceptrón

### Entrenamiento de la Red

Así como entrenamos nuestro cerebro para hacer una cierta tarea (lo llamamos aprendizaje), necesitamos entrenar una red cruda para realizar tareas. Este entrenamiento requiere enormes recursos computacionales, dependiendo del tamaño y la complejidad de la red. Así como le enseñas repetidamente al niño lo mismo hasta que lo aprenda completamente, *ANN* también aprende a través de varias rondas de entrenamiento, llamadas iteraciones.

El aprendizaje mejora a través de muchas iteraciones (*epochs*). Necesitas mostrar información nueva y adicional a la red en cada iteración. Por lo tanto, la cantidad de datos que se requiere para entrenar una red típica suele ser muy grande: varios miles de puntos de datos. Como científico de datos, este es un punto importante a tener en cuenta: podemos entrenar un algoritmo clásico con solo unos cientos de puntos de datos, mientras que una red neuronal típicamente requiere miles de instancias. Con el concepto de *big data*, fácilmente tenemos millones de puntos de datos disponibles para *ANN*.

### Arquitecturas de ANN

Hay principalmente tres tipos de arquitecturas:

- Red de una sola capa sin retroalimentación  
- Red de múltiples capas sin retroalimentación  
- Red recurrente  

Una red de una sola capa sin retroalimentación, también llamada perceptrón de una sola capa, contiene solo las neuronas de cómputo de la capa de salida. Como no realizamos cómputo en la capa de entrada, nunca se cuenta como una capa. Esta es una red acíclica. En redes de múltiples capas, hay una o más capas ocultas entre la capa de entrada y la de salida. La capa de entrada suministra la señal de entrada a la primera capa oculta, cuya salida va como entrada a la segunda, y así sucesivamente. La salida de la última capa se trata como tu salida final.

En el caso de una red recurrente, es como una red sin retroalimentación, excepto que tiene al menos un bucle de retroalimentación desde una de las capas ocultas hacia una capa oculta anterior. Con esta comprensión básica de *ANN*, ahora veamos sus avances, es decir, *DNN* y los diferentes tipos de redes.

### ¿Qué es DNN?

Cuando la arquitectura de la red se vuelve realmente compleja con muchas capas, cada una con muchos nodos, la llamamos red neuronal profunda (*DNN*). No hay una clara demarcación entre *ANN* y *DNN*, y muchas veces la gente usa esos términos indistintamente. No hay nada de malo si tú también lo haces.

Nuestro cerebro puede aprender múltiples cosas y recordarlas. Sin embargo, las *DNNs* diseñadas hasta ahora solo pueden realizar un tipo de tarea. Por lo tanto, tenemos diferentes arquitecturas *DNN*, cada una diseñada específicamente para una cierta tarea. Después de que una red es entrenada, realizará solo una tarea específica para la que fue entrenada. Por ejemplo, una cierta red puede hacer traducción de idiomas, mientras que entrenamos otra para detección de rostros, y así sucesivamente. Diferentes redes para diferentes tareas.

Tenemos un rol separado para diseñar la red. Lo llamamos ingeniero de *machine learning* (*ML*). El ingeniero de *ML*, al diseñar la red, decide el número de capas, el número de nodos en cada capa, la topología de la red, la función de activación, el algoritmo de optimización, y así sucesivamente. ### Funciones de Activación

Una función de activación en una red neuronal define cómo se transforma la suma ponderada de la entrada a un nodo en una salida. Decide si una neurona se activa o no. Introduce no linealidad en las operaciones realizadas por las diversas funciones en toda la red. Has visto que la salida de cada neurona está dada por la ecuación $W \cdot X$; sin una función de activación, simplemente apilaría las salidas de todas las capas unas sobre otras, resultando en solo una combinación lineal de todas. Por lo tanto, sin una función de activación, toda la red es esencialmente un modelo de regresión lineal. La función de activación ayuda a mantener los valores de salida restringidos a un cierto rango.

Dependiendo del tipo de no linealidad que quieras introducir en la ecuación, los diseñadores crearon varios tipos de funciones de activación.

Aquí hay algunas funciones de activación ampliamente utilizadas:

- *Sigmoid*: tradicionalmente utilizada para clasificación binaria; si $c \leq 0.5$ entonces $y=0$ o de lo contrario $y=1$. Sufre de problemas de desvanecimiento del gradiente y es computacionalmente costosa. Si los gradientes se acercan a 0 o 1, la tasa de aprendizaje se vuelve demasiado lenta.
- *Tanh*: también se usa para clasificación binaria; si $x \leq 0, y=0$ de lo contrario $y=1$. A diferencia de *sigmoid*, esta está centrada en cero, lo que significa que los valores varían entre $-1$ y $+1$. También es computacionalmente costosa.
- *ReLU*: Unidad Lineal Rectificada; si $x \leq 0$, $y=0$ de lo contrario $y=x$. Por lo tanto, se activa solo para valores positivos. Es computacionalmente menos costosa en comparación con *sigmoid* y *tanh*.
- *Leaky-ReLU*: Resuelve el problema de *ReLU muerta* agregando una pequeña pendiente positiva en el plano negativo. Ha sido ampliamente utilizada en *GANs* (redes generativas antagónicas—aprenderás más sobre *GANs* en breve).
- *Softmax*: Esta función nos ayuda a representar la entrada en términos de una distribución de probabilidad discreta. Aplicamos una función exponencial a cada elemento de la capa de salida y normalizamos sus valores para asegurar que su suma sea igual a 1. Consideramos la clase con la puntuación de confianza más alta como la salida final. Principalmente la usamos como la última capa en problemas de clasificación multiclase.

#### Rectified Linear Unit (ReLU)

```python
def RELU(x):
    ''' It returns zero if the input is less than zero otherwise it returns the given input. '''
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1
```

```python
x = np.linspace(-10, 10)
plt.plot(x, RELU(x))
plt.title('Activation Function :RELU')
plt.axhline(y=0, color='0', linestyle='-')
plt.axvline(x=0, color='0', linestyle='-')
plt.grid()
```

#### Lecky Relu

```python
def Lecky_relu(x,a=0.01):
  x2=[]
  for i in x:
    if i<0:
      i=a*i
    x2.append(i)
  return x2  
```

```python
x = np.linspace(-10,10)
plt.plot(x, Lecky_relu(x))
plt.title('Activation Function :Lecky Relu')
plt.axhline(y=0, color='0', linestyle='-')
plt.axvline(x=0, color='0', linestyle='-')
plt.grid()
```

#### Parametric Rectified Linear Unit (PReLU)

```python
def Param_relu(x,a=0.1):
  x2=[]
  for i in x:
    if i<0:
      i=a*i
    x2.append(i)
  return x2  
```

```python
x = np.linspace(-10, 10)
y= Param_relu(x,a=0.1)
plt.plot(x, y)
plt.title('Activation Function :Parametric ReLU')
plt.axhline(y=0, color='0', linestyle='-')
plt.axvline(x=0, color='0', linestyle='-')
plt.grid()
```

#### Exponential Linear Unit (ELU)

```python
def ELU(x,a=1.0):
  x3=[]
  for i in x:
    if i<0:
      i=a*(np.exp(i)-1)
    x3.append(i)
  return x3 
```

```python
x = np.linspace(-10, 10)
y= ELU(x,a=1.0)
plt.plot(x, y)
plt.title('Activation Function :Parametric ReLU')
plt.axhline(y=0, color='0', linestyle='-')
plt.axvline(x=0, color='0', linestyle='-')
plt.grid()
```

#### Sigmoid

```python
x = np.linspace(-10, 10)
y = 1/(1 + np.exp(-x))
plt.plot(x,y)
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.axhline(y=0, color='0', linestyle='-')
plt.axvline(x=0, color='0', linestyle='-')
plt.grid()
```

#### Softmax

```python
def softmax(x):
    """ applies softmax to an input x"""
    e_x = np.exp(x)
    return e_x / e_x.sum()
```

```python
x = np.linspace(-10, 10)
y = softmax(x)
plt.plot(x,y)
plt.axis('tight')
plt.title('Activation Function :Softmax')
plt.axhline(y=0, color='0', linestyle='-')
plt.axvline(x=0, color='0', linestyle='-')
plt.grid()
```

### Retropropagación

Cuando diseñas una red de múltiples capas, asignas algunos pesos iniciales a los diversos nodos en cada capa. Dependiendo de la salida final, ajustas los valores de estos pesos y pruebas la salida nuevamente. Repites este proceso múltiples veces (llamadas *epochs*) hasta que estés satisfecho con tu salida. Ahora, ¿cómo sabes en qué dirección ajustar los pesos? Me refiero a si aumentar o disminuir sus valores. Aquí viene el concepto de optimización y funciones de pérdida. Utilizas alguna función de optimización que ayudará a reducir la diferencia (error) entre la salida actual (predicha) y la salida deseada (real). En cada iteración, intentas minimizar la función de pérdida. Te mueves hacia atrás desde la capa de salida hasta la capa de entrada, haciendo correcciones a los pesos en cada capa visitada. A esto lo llamamos *retropropagación*. La función de optimización debe ser diferenciable para que puedas determinar los mínimos. Y es por eso que he incluido las derivadas de las funciones en la última columna de la tabla de funciones de activación.

Los valores de salida pueden volverse infinitesimalmente pequeños o enormes debido a las operaciones de optimización realizadas en cada capa. Esto puede causar los conocidos problemas de gradiente desvanecido y explosivo. Entonces, ¿qué son los gradientes desvanecidos y explosivos?

### Gradientes Desvanecidos y Explosivos

Como ahora entiendes, durante el entrenamiento, los gradientes se retropropagan hasta la capa inicial. En cada capa, hacemos multiplicaciones de matrices. Para cadenas largas (gran número de capas), el gradiente seguirá reduciéndose exponencialmente a valores muy bajos y eventualmente desaparecerá, en el sentido de que la red dejará de aprender. A esto lo llamamos el problema del gradiente desvanecido. Por otro lado, si el gradiente se mueve en la otra dirección tomando valores grandes, explotará en algún momento y colapsará el entrenamiento. A esto lo llamamos el gradiente explosivo.

A continuación, explicaré la optimización.

### Funciones de Optimización

El algoritmo de optimización es el aspecto más importante del entrenamiento de una red neuronal. Una función de pérdida, también llamada función de error, es la función objetivo para la optimización. El algoritmo de optimización minimiza la función de pérdida y, por lo tanto, a veces se le denomina función de minimización. Dependiendo del objetivo, usamos el optimizador para minimizar o maximizar la función objetivo. No hace falta decir que la función de pérdida necesita ser diferenciable. Por lo tanto, podemos decir que la optimización es un proceso de fuerza bruta para ajustar parámetros en la función de pérdida para lograr la varianza mínima entre los valores predichos y los reales. En nuestro caso de *ANN*, el peso y el sesgo se convierten en los parámetros que necesitan ajustes, y la función de pérdida es alguna función definida por nosotros con la condición de que debe ser continua y diferenciable.

En el entrenamiento de redes neuronales, principalmente usamos el descenso de gradiente como estrategia de optimización. El optimizador de descenso de gradiente intenta alcanzar los valores óptimos para los pesos asignados a cada nodo en cada capa para los cuales el gradiente de la función de pérdida es mínimo. El gradiente no es más que la tasa de cambio en el valor de alguna variable. En nuestro caso, es la tasa a la que cambia la función de pérdida con respecto a los pesos. ### Tipos de Optimizadores

A continuación, doy una lista de optimizadores comúnmente utilizados junto con su breve descripción. Como científico de datos, esto debería ser suficiente para diseñar tus propias arquitecturas *ANN*. No necesitas intentar entender las matemáticas detrás de estos; solo el conocimiento del funcionamiento conceptual sería suficiente.

- **Descenso de gradiente simple**: Calcula el gradiente para todo el conjunto de datos de entrenamiento. Aunque idealmente esto es lo correcto que se debe hacer, el algoritmo es computacionalmente costoso.
- **Descenso de gradiente estocástico (*SGD*)**: Calcula el gradiente y actualiza su peso para cada par $(x_i, y_i)$. Esto es comparativamente más rápido que el descenso de gradiente simple y también es computacionalmente menos costoso. También tiene otra enorme ventaja: que puede actualizar pesos con nuevos puntos de datos agregados sobre la marcha.
- **Descenso de gradiente por mini lotes**: Como sugiere el nombre, el algoritmo realiza las actualizaciones en un lote de puntos aleatorios en el conjunto de datos. Esto proporciona actualizaciones más estables en los pesos en comparación con el descenso de gradiente estocástico. También es computacionalmente menos costoso. Enviamos el tamaño del lote como un parámetro al algoritmo. Muchas bibliotecas modernas de aprendizaje profundo utilizan este algoritmo.
- **Momento**: En el caso de *SGD* y mini lotes, los investigadores observaron fluctuaciones y notaron que no sigue un camino recto y constante hacia el óptimo. Para eliminar este ruido, los investigadores introdujeron el momento en la función de actualización. Las fluctuaciones se redujeron promediando los parámetros pasados.
- **Adagrad**: En este algoritmo, el gradiente es adaptativo en el sentido de que para iteraciones más altas, se reduce la tasa de aprendizaje.
- **RMSProp**: Este algoritmo fue desarrollado para superar las deficiencias de *Adagrad*. No decae la tasa de aprendizaje demasiado rápido. *RMSProp* significa *Root Mean Square Propagation* y, en la mayoría de los casos, supera a *Adagrad*.
- **AdaDelta**: Esto significa *Adaptive Delta* y elimina las deficiencias en *Adagrad*. El delta es la diferencia entre el peso actual y el peso recién actualizado. El algoritmo utiliza el promedio móvil exponencial (*EMA*) de los cuadrados de los deltas en lugar de la tasa de aprendizaje. Por lo tanto, no hay necesidad de anotar explícitamente la tasa de aprendizaje.
- **Adam**: *Adam* utiliza lo mejor de ambos mundos, momento y *AdaDelta*. Su nombre completo es *Adaptive Moment Estimation*. Utiliza tanto el promedio exponencial decreciente de gradientes cuadrados pasados como en *AdaDelta* y también de gradientes pasados como en momento.

### Funciones de Pérdida

Para evaluar el entrenamiento del modelo, usamos las funciones de pérdida. Su valor nos da una medida de la precisión de nuestro entrenamiento: qué tan lejos está el valor predicho del real. Cuanto más lejos, más entrenamiento se requiere; esencialmente, significa entrenar por más *epochs*. Usamos diferentes funciones de pérdida para regresión y clasificación.

#### Funciones de Pérdida para Regresión

Describiré tres funciones de pérdida para regresión ampliamente utilizadas:

- **MAE**: error absoluto medio; también llamado pérdida $L_1$. El error absoluto medio es el promedio de las diferencias absolutas entre los valores predichos y reales. Hay un problema con esta función de pérdida: los valores subestimados altamente cancelan los valores sobrestimados, dándonos una idea equivocada del error neto.
- **MSE**: error cuadrático medio, también llamado pérdida $L_2$. Aquí, tomamos el promedio de las diferencias al cuadrado entre los valores predichos y reales como nuestro término de error. Elevar al cuadrado el error nos ayuda a resolver los problemas de *MAE*. Además, cuando los valores de entrada y salida tienen escalas pequeñas, el *MSE* resaltará mejor el error debido al cuadrado.
- **Huber**: también llamado error absoluto medio suavizado. Es menos sensible a valores atípicos que *MSE*. Da el error absoluto. Puede amplificar errores menores a cuadráticos. El hiperparámetro llamado $\delta$ decide esto. La pérdida de *Huber* se acerca a *MSE* cuando $\delta$ se aproxima a 0 y a *MAE* cuando $\delta$ se aproxima a infinito.

#### Funciones de Perdida para Clasificación

- **Entropía cruzada binaria**: comúnmente utilizada para clasificación binaria.
- **Hinge**: otra función de entropía para clasificación binaria. Principalmente desarrollada para *SVMs*. Usamos esto cuando los valores objetivo están dentro del conjunto de deato
- **Entropía cruzada multiclase**: utilizada para clasificaciones multiclase. La entropía cruzada calcula puntuaciones para cada clase que significan la diferencia entre los valores predichos y reales.
- **Entropía cruzada multiclase dispersa**: en este caso, la salida está en forma de vectores codificados *one-hot* en lugar de los valores enteros.

## Red Neuronal Convolucional

Principalmente diseñada para resolver problemas de visión por computadora, como clasificación de imágenes y reconocimiento de objetos. La red comprende tres tipos de capas:

- **Convolucional**
- ***Pooling***
- **Totalmente conectada (*FC*)**

La capa convolucional es la primera capa, y *FC* es la última capa. Entre medias, contiene más capas convolucionales y de *pooling*. Cada capa convolucional identifica algunas características en la imagen, aumentando progresivamente en complejidad hasta que toda la red finalmente identifica el objeto pretendido. La red utiliza fuertemente multiplicaciones de matrices mientras identifica patrones dentro de una imagen y, por lo tanto, es computacionalmente costoso.

### Capa Convolucional

Este es el bloque de construcción principal de *CNN*. Comprende tres componentes: datos de entrada, núcleo (*kernel*) y el mapa de características, como se muestra en la figura 17.6.

  <img src="content/fig17.6.png" width="350"/>
</div>

<p style="color:blue; margin-top: 5px; font-weight: bold;">Fig. 17.6</p>

<p>Convoluciones.</p>

Movemos el núcleo o filtro a través de los campos de una imagen para detectar características. Este proceso se llama convolución. El núcleo es una matriz bidimensional de pesos y típicamente tiene un tamaño de una matriz de $3 \times 3$. Calculamos el producto punto entre los píxeles de entrada y el filtro. Luego, el filtro se desplaza por un *stride*, repitiendo el proceso hasta que recorra toda la imagen. Llamamos al arreglo de salida de todos estos productos punto un mapa de características, mapa de activación o característica convolucionada. Como el arreglo de salida no se mapea directamente a los valores de entrada, consideramos esta capa como una capa parcialmente conectada. Necesitas decidir los valores de tres hiperparámetros antes de que comience el entrenamiento:

- Número de filtros que deciden la profundidad de salida.
- *Stride* que decide el número de píxeles que se mueve el núcleo. Cuanto mayor sea el *stride*, más pequeña será la salida.
- *Padding*: esto decide cómo se rellena la salida cuando el filtro no encaja en la imagen de entrada.

A la salida de una capa *CNN*, usamos *ReLU* para introducir la no linealidad. Apilamos muchas de estas capas convolucionales en una jerarquía lineal. A la salida de la última capa, obtenemos los valores numéricos para los diversos patrones identificados en la imagen.

La capa de *pooling* se utiliza para la reducción de dimensionalidad. El *pooling* puede ser de dos tipos: máximo o promedio. Se puede observar en la figura 17.7

El *max pooling* toma el valor máximo de un píxel cubierto por el núcleo, mientras que el *average pooling* calcula el promedio de todos los píxeles cubiertos por el núcleo. Aunque se pierde mucha información debido al *pooling*, ayuda a reducir la complejidad, mejorar la eficiencia y limitar el riesgo de sobreajuste.

### Capa Totalmente Conectada

Esta capa realiza la tarea de clasificación basándose en las características extraídas de toda la jerarquía *CNN*. Utiliza activación *softmax*, dando valores de probabilidad en el rango de 0 a 1, a cada clase.

### Aplicaciones de *CNN*

Se han desarrollado varias variantes de *CNN*, por nombrar algunas: *AlexNet*, *VGGNet*, *GoogleNet*, *ResNet*, y así sucesivamente. Todas ellas realizan tareas de reconocimiento de imágenes y visión por computadora para derivar información significativa de imágenes y videos. Basándote en esta información, puedes iniciar acciones apropiadas. Debido a esta capacidad de recomendación, *CNN* se ha aplicado en muchas áreas. Las redes sociales la utilizan para etiquetar personas en una fotografía. Los médicos la usan para identificar tumores malignos. Las empresas minoristas la utilizan para recomendar artículos similares basándose en los artículos comprados (identificados) anteriormente. Finalmente, también puede aplicarse a autos sin conductor para detectar carriles y señales de tráfico.

### Arquitectura del Modelo

*GAN* comprende dos redes neuronales, ambas basadas en *CNN*. Llamamos a una generadora y a la otra discriminadora. Un proceso antagónico entrena simultáneamente ambas redes. El generador aprende a crear imágenes que parecen reales. El discriminador, que es un crítico, se entrena para distinguir una imagen real de las falsas. Si un generador supera al discriminador, tendríamos la capacidad de crear imágenes que parecen reales.

El entrenamiento de *GAN* se realiza en dos partes: 

En la primera parte, mantenemos el generador inactivo y entrenamos solo al discriminador. Cuando el discriminador no puede decir que la imagen generada es falsa, detenemos el entrenamiento. Ahora veamos qué uso tiene una *GAN* para un científico de datos.

Los conjuntos de datos de imágenes siempre son escasos en términos de sus puntos de datos. Has utilizado el conjunto de datos *MNIST* para dígitos manuscritos y también alfabetos. La aplicación era detectar actividades maliciosas (ataques) en una red. Obtener los datos de red para entrenamiento y pruebas era un desafío enorme. ## Redes Neuronales Recurrentes (*RNN*)

Al igual que *CNN*, *RNN* fue investigada hace décadas; sin embargo, su potencial completo no pudo ser explotado debido a la falta de recursos computacionales en aquellos días. Con la disponibilidad de recursos, utilizaron exitosamente *RNN* en una amplia variedad de dominios. La usamos en voz a texto, traducciones de idiomas, pronósticos de series temporales, subtítulos de imágenes, preguntas y respuestas, y muchas otras aplicaciones.

El punto clave a tener en cuenta en *RNN* es su capacidad de recordar el pasado. Una red neuronal densa típica no tiene la capacidad de recordar lo que aprendió en el pasado. En áreas como las traducciones de idiomas, recordar las palabras anteriores es importante; de lo contrario, la traducción se convertiría en una mera traducción palabra por palabra, que tendría poco sentido en el idioma traducido. Solo para tu conocimiento rápido, la figura 17.11 muestra la arquitectura *RNN*.

Como se ve en el diagrama, cada nodo, excepto el primero, recibe la salida del nodo anterior como entrada adicional. Por lo tanto, la salida de cada nodo dependerá de su propia entrada, más los hallazgos del nodo anterior. Así, los hallazgos pueden propagarse a través de muchas capas hasta la salida final. Cuando entrenas *RNNs* para profundidades mayores, encontrarás el problema del gradiente desvanecido. Debido a esta limitación de *RNN*, pronto fueron reemplazadas por otras arquitecturas mejoradas.
Como científico de datos, no necesitas profundizar en el funcionamiento de *RNN*. El modelo no está en uso común en estos días. Sin embargo, la comprensión básica y el concepto de recordar el pasado es importante para entender las arquitecturas posteriores.

$$
f_t = \sigma(W_f \cdot [h_{t-1}, x_t] + b_f)
$$

La *input gate* toma como entrada la entrada actual y el estado oculto anterior. La capa de entrada *sigmoid* decide qué valores actualizar, y la activación *tanh* crea un vector de nuevos valores candidatos. Las dos salidas, *sigmoid* y *tanh*, luego se multiplican. La salida contiene la información importante del módulo anterior: la marca de tiempo. Matemáticamente, esto se expresa como:

$$
i_t = \sigma(W_i \cdot [h_{t-1}, x_t] + b_i) \\
\tilde{C}_t = \tanh(W_C \cdot [h_{t-1}, x_t] + b_C)
$$

$$

$$

Las redes neuronales profundas suelen tener arquitecturas muy complejas, requieren recursos pesados para entrenamiento, toman semanas de tiempo de entrenamiento y consumen muchos datos en tamaños de varios gigabytes incluso terabytes. ### Modelos Preentrenados para Texto

Te presentaré brevemente los siguientes cinco modelos preentrenados:

- *Word2Vec*
- *GloVe*
- *Transformers*
- *BERT*
- *GPT*

Cada uno fue diseñado para un propósito específico y encuentra su uso en varias aplicaciones de *ML* que usan conjuntos de datos de texto.

#### *Word2Vec*

Considera desarrollar un modelo de *ML* para predecir la siguiente palabra o encontrar una palabra faltante en una oración. Desarrollar dichos modelos requiere un buen vocabulario, que ayuda a capturar el contexto de una palabra en un documento, descubriendo similitudes semánticas/sintácticas entre palabras, relación con otras palabras en el documento, y así sucesivamente. Como puedes imaginar, un vocabulario puede ser extremadamente grande incluso para un documento pequeño a mediano. Representar todo el vocabulario en un formato legible por máquina, en formato binario, requeriría una memoria prodigiosa. Los *embeddings* proporcionan una representación compacta del vocabulario. *Word2Vec* es un método para construir dicho *embedding*. El *embedding* puede ser construido usando dos métodos: *Skip Gram* y *CBOW* (Common Bag Of Words). Muchos modelos *Word2Vec* preentrenados de gran tamaño están disponibles para desarrollar diversos tipos de aplicaciones de *NLP*. Incluso encontrarás modelos *Sentence2Vec* que computan oraciones...

#### *GloVe*

*GloVe*, un proyecto de código abierto lanzado en 2014 en *Stanford*, es un algoritmo de aprendizaje automático no supervisado que crea representaciones vectoriales para palabras. El algoritmo descubre las similitudes semánticas entre las palabras dando razones de probabilidades de coocurrencia palabra a palabra. Por ejemplo, la probabilidad de que la palabra "cream" ocurra junto a "ice" sería muy alta. El algoritmo fue aplicado en un enorme corpus de texto como *Wikipedia*, *Common Crawl* y *Twitter*, y se desarrollaron los *embeddings* para dimensiones de 50, 100, 200 y 300. Puedes usar estos *embeddings* en tus proyectos de *NLP* para construir tus propios directivos.

#### *Transformer*

Esta es la innovación más reciente en modelado de lenguaje natural. Virtualmente ha reemplazado a *LSTMs*: el modelo utilizado popularmente para desarrollar aplicaciones de *NLP*. Está basado en el famoso artículo *"Attention Is All You Need"* y ha mejorado drásticamente el rendimiento de la traducción automática neuronal.

#### *BERT*

*BERT* significa *Bidirectional Encoder Representations from Transformers*. Jacob Delvin et al. introdujeron este modelo en 2018 en su artículo titulado *"BERT: Pre-training of Deep Bidirectional Transformers for Language Understanding"*. Puedes usar fácilmente un modelo *BERT* preentrenado en tus aplicaciones de *NLP*: solo agregar una capa es todo lo que se necesita para hacerlo. Puedes crear fácilmente aplicaciones altamente precisas de traducción de idiomas, preguntas y respuestas, y así.

#### *GPT*

Este es un modelo de lenguaje autoregresivo que produce texto similar al humano. Desarrollado en *OpenAI*, ha dado un rendimiento sobresaliente en generación de texto. Es tan poderoso en generación de texto que uno puede crear fácilmente noticias falsas que pueden pasar por reales. Debido a esto y otras aplicaciones maliciosas que podrían crearse usando *GPT*, *OpenAI* ha lanzado solo un modelo pequeño para experimentar.

A diferencia de *BERT*, *GPT* es unidireccional y fue entrenado en 175 mil millones de parámetros. La versión actual al momento de escribir esto es *GPT-3*, y se lanzan nuevas versiones periódicamente.

Con esto, pasemos a los modelos preentrenados desarrollados para datos basados en imagen.

### Modelos Preentrenados para Datos de Imagen

El pionero en este campo es probablemente *AlexNet*, una arquitectura *CNN* que logró una posición entre los 5 primeros en el *ImageNet Large Scale Visual Recognition Challenge* en septiembre de 2012. *ImageNet*, el primero de su tipo en términos de escala, comprende más de 14 millones de imágenes etiquetadas en 27 subárboles y 21,841 subcategorías. Desarrollaron muchos modelos de aprendizaje automático para clasificar este conjunto de datos. Dos años después, *Google DeepMind* y la Universidad de Oxford lanzaron *VGG16* y *VGG19*. Siguiendo esto, tenemos una familia de arquitecturas en esta categoría incluyendo *ResNet*, *Inception*, *Xception*, *MobileNet*, *DenseNet*, *EfficientNet*, e híbridos. Te presentaré brevemente algunos modelos en esta categoría.

*ResNet* (Red Residual), introducida en 2015 por Shaoqing et al., es uno de los modelos de aprendizaje profundo populares y exitosos. Fue construida sobre constructos tomados de células piramidales en la corteza cerebral. Comprende 50, 101 o 151 capas. Utilizan conexiones de salto para saltar sobre algunas capas. Proporciona una alta precisión en comparación con otras redes en su categoría. Toma mucho tiempo entrenar.

*VGG16* es un modelo *CNN* de investigadores de la Universidad de Oxford que logró un 92.7% de precisión entre los 5 primeros en *ImageNet*. Lo entrenaron durante varias semanas usando *NVIDIA Titan Black* GPUs.

*Inception V3* alcanzó más del 78.1% de precisión en *ImageNet*. *Xception* superó ligeramente a *Inception V3* en *ImageNet*, pero mostró una mejora significativa sobre *Inception V3* en conjuntos de datos más grandes que comprenden 350 millones de imágenes y 17,000 clases. *DenseNet* es otra arquitectura *CNN* que alcanzó resultados de vanguardia en *ImageNet*. *EfficientNet* mejora la precisión y eficiencia utilizando *AutoML* y empleando escalamiento de modelado.

*MobileNet* fue diseñada teniendo en mente dispositivos móviles para clasificación de imágenes, detección de objetos, y más.

*YOLO* (*You Only Look Once*) es un sistema de detección de objetos en tiempo real de vanguardia que dibuja cuadros delimitadores alrededor de los objetos detectados y predice su clase con las probabilidades correspondientes. En realidad, es un algoritmo que usa *ANN* para realizar detección de objetos en tiempo real. Es bastante rápido y preciso. Lo usamos en una amplia variedad de aplicaciones como detectar señales de tráfico, parquímetros, y así sucesivamente. Si te encuentras con un requisito para analizar el tráfico, por ejemplo, cuántos vehículos pesados se dirigen hacia el este en una intersección particular, piensa en usar *YOLO*.

Finalmente, como científico de datos, tienes varias opciones de modelos preentrenados para desarrollar tus modelos de *ML* basados en imagen.

## Ventajas/Desventajas

Aquí hay algunas de las ventajas de usar la tecnología *ANN*:

- Las redes neuronales artificiales pueden aprender fácilmente incluso relaciones no lineales y complejas entre la entrada y la salida.
- Puede manejar conjuntos de datos enormes, lo cual es una limitación severa de *GOFAI*.
- Podemos distribuir el entrenamiento a través de procesadores paralelos, lo que reduce los tiempos de entrenamiento.
- Una vez entrenada, también tiene la capacidad de generalizar, es decir, puede inferir nuevas relaciones no vistas.
- Una ventaja importante es que la ingeniería de características está integrada. Durante el entrenamiento, las características se extraen automáticamente.

Aquí hay algunas desventajas:

- Computacionalmente más costosa que *GOFAI*.
- Requiere conjuntos de datos mucho más grandes en comparación con los requeridos en el caso de *ML* clásico.
- El diseño de la red es crítico y puede tomar mucho tiempo lograr los resultados deseados. (En estos días tienes herramientas similares a *AutoML* que ayudan a diseñar redes eficientes.)
- Explicabilidad: una red neuronal funciona como una caja negra. No sabes por qué llegó a cierta salida.

## Resumen

La introducción de la tecnología *ANN* definitivamente ha traído un cambio radical en el aprendizaje automático. Aunque la tecnología tiene muchas décadas de antigüedad, su éxito llegó solo con la disponibilidad de recursos computacionales que tenemos hoy. Mientras desarrollan los modelos de aprendizaje automático, los científicos de datos ahora tienen una elección entre las dos tecnologías: basada en estadísticas clásicas versus basada en redes neuronales artificiales. La tecnología *ANN/DNN* requiere recursos pesados, conjuntos de datos enormes y varias horas/días/semanas de tiempos de entrenamiento, mientras que la tecnología tradicional clásica de *ML* funciona superbamente para conjuntos de datos pequeños. La tecnología *ANN* falla totalmente si los conjuntos de datos son demasiado pequeños. Así que, esto puede convertirse en un factor decisivo al seleccionar la tecnología para el desarrollo del modelo.
