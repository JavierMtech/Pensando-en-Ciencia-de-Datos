# Clustering: Visión General

El *clustering* es un aspecto importante del análisis de datos y el aprendizaje no supervisado más importante. Su objetivo es dividir un conjunto de datos en subconjuntos. Los objetos en el mismo subconjunto son similares entre sí respecto a cierta medida de similitud. Los objetos en diferentes *clusters* son disímiles según la misma medida. Usas *clustering* para entender los datos o como un primer paso en operaciones posteriores, como indexación y compresión de datos. La necesidad de entender los datos surge en reconocimiento de patrones, minería de datos, recuperación de información, etc.  

El proceso de *clustering* generalmente sigue estos pasos:  
• Extraer y seleccionar las características más representativas.  
• Diseñar un algoritmo de *clustering* nuevo o seleccionar uno conocido.  
• Evaluar los resultados para juzgar la validez del algoritmo.  
• Explicar los resultados a tu cliente. Si el cliente está satisfecho con tu explicación, tu trabajo ha terminado.

Como la noción de “*cluster*” no está definida de manera precisa, los investigadores desarrollaron muchos modelos de *clustering* y, por ende, muchos algoritmos. Entender estos modelos es importante para aprender los algoritmos. La agrupación jerárquica construye modelos llamados de conectividad basados en la distancia. Los modelos de *centroides*, como el algoritmo *K-means*, forman *clusters* mediante un único vector promedio. Los modelos de distribución agrupan datos según su distribución estadística. Usamos el algoritmo de *expectation-maximization* (*EM*) para encontrar *clusters* en distribuciones multivariantes. Luego están los modelos basados en grafos como *CLIQUE*. Algunos algoritmos, como *DBSCAN*, solo proporcionan información de agrupamiento para subgrupos. El *grid-based clustering*, como *STING*, divide el conjunto de datos en subespacios. Incluso hay *clustering* estilo *gossip*, como *affinity propagation*, que usa mensajería entre pares para formar grupos o *clusters*. También hay modelos neuronales como el *self-organizing map*, que es una red neuronal no supervisada.  

Como científico de datos, necesitarás aplicar uno o más de estos modelos para proporcionar análisis significativos a tu cliente. Los algoritmos de *clustering* se dividen ampliamente en estas categorías:  
- Basado en centroides  
- Basado en conectividad  
- Basado en distribución  
- Basado en densidad  
- Clustering de grandes conjuntos de datos  
- Clustering estilo gossip  
- Basado en grillas  

En el *clustering* basado en centroides, partimos de la suposición de que los datos contienen $K$ *clusters*. Los *clusters* se forman resolviendo el problema de optimización definido como encontrar $K$ centros de *cluster* y asignar los objetos al centro más cercano. Una función de distancia mide la cercanía de los puntos para determinar el centro más próximo. El problema de optimización es NP-hard, y por lo tanto esta técnica solo puede ofrecer soluciones aproximadas. El *K-means clustering* cae en esta categoría. *K-medoids*, *K-medians* y *K-means++* son algunas variaciones. La mayor desventaja de este algoritmo es que requiere estimar el número de *clusters* de antemano. También se asume que los *clusters* tienen aproximadamente el mismo tamaño. Como el algoritmo optimiza los centros de los *clusters* y no los bordes, se obtienen bordes cortados incorrectamente. El *mean shift clustering* es otro algoritmo que entra en esta categoría. Este algoritmo descubre los *clusters* en una densidad suave de puntos de datos al desplazar iterativamente el *kernel* hacia una región de mayor densidad hasta la convergencia.  

En el *clustering* basado en conectividad, conectas objetos para formar *clusters* según su distancia. Se asume que los objetos cercanos están más relacionados entre sí que los lejanos. Una distancia máxima decide un *cluster*. Puedes representar toda la estructura como un dendrograma, y por eso este tipo de *clustering* también se llama jerárquico. Al recorrer el dendrograma, los *clusters* se fusionan entre sí a ciertas distancias. El *clustering* depende no solo de la elección de la función de distancia sino también del criterio de enlace (*linkage*). Puede ser un enlace simple, completo o promedio. El *clustering* puede ser aglomerativo o divisivo. En el aglomerativo, comienzas con un solo punto de datos y agregas otros puntos para formar un *cluster*. En el divisivo, consideras todo el conjunto de datos como un solo *cluster* y luego lo divides en particiones. Por esta descripción, puedes entender que estos métodos no producen una partición única de los datos, sino que crean una jerarquía de la que el usuario debe elegir los *clusters* apropiados.  

El *clustering* basado en distribución se basa en modelos de distribución estadística. Los objetos que pertenecen a la misma distribución forman un *cluster*. Una distribución típica es la gaussiana; asumimos que todo el conjunto de datos comprende *clusters*, cada uno con una distribución gaussiana de puntos dentro de él. Los *Gaussian mixture models* son una de estas técnicas de *clustering*. Este tipo de *clustering* es bueno para capturar correlación y dependencia entre atributos. Sin embargo, la suposición en la que se basa (distribución gaussiana) es bastante fuerte y, por lo tanto, puede no producir un buen *clustering* si los puntos de datos no siguen la distribución esperada. 

En el *clustering* basado en densidad, una alta densidad de puntos en un área determinada define el *cluster*. Una caída en la densidad marca los límites del *cluster*. Tales algoritmos funcionan bien para distribuciones como mezclas gaussianas. El algoritmo *DBSCAN* cae en esta categoría. Al igual que el *linkage-based clustering*, agrupamos los puntos dentro de un umbral de distancia determinado con la condición adicional de que haya un número mínimo de otros objetos dentro del mismo radio. Como el *cluster* comprende todos los objetos conectados por densidad, puede crear *clusters* de cualquier forma arbitraria. La complejidad de este algoritmo es bastante baja y las salidas son determinísticas, produciendo los mismos resultados en cada ejecución. *OPTICS* es otro algoritmo que pertenece a esta categoría. Generaliza *DBSCAN* al eliminar la necesidad de cambiar el valor del parámetro de rango ε. Produce una visualización de los enlaces entre los puntos; el usuario crea los *clusters* de su elección usando estas visualizaciones. El *mean shift clustering* es otro algoritmo de esta categoría, en el que cada punto de datos se mueve hacia la zona más densa más cercana basada en la estimación de densidad del *kernel*. El *mean shift* puede detectar *clusters* de forma arbitraria. Generalmente es más lento que *DBSCAN* y no produce buen *clustering* en conjuntos de datos multidimensionales. 

¿Cómo agrupas un conjunto de datos realmente enorme? Usas la estrategia *divide-and-conquer* (dividir y conquistar). Para conjuntos de datos muy grandes, lo divides en conjuntos más pequeños, conservando la mayor cantidad posible de información del original en los subconjuntos. Luego agrupas cada unidad compacta usando otras técnicas de *clustering* conocidas. Así funciona el algoritmo *BIRCH*. *CLARANS* es otro algoritmo que usamos para *clustering* de grandes conjuntos de datos. Divide el conjunto de datos grande mediante una búsqueda aleatoria en unidades más pequeñas, luego aplica el algoritmo *PAM* (*partitioning around medoids*) en cada subconjunto, determina la calidad del *clustering* midiendo las disimilitudes y repite todo el proceso de muestreo y *clustering* hasta que te satisfaga el resultado.  

Luego viene otra categoría de mensajería, que es estilo *gossip*. La llamamos *affinity propagation clustering*, donde formamos los grupos o *clusters* mediante *gossip*, es decir, mensajería entre pares. 

Todos los algoritmos anteriores son basados en consultas, donde los resultados intermedios de una consulta no se guardan. Tienes que escanear todo el conjunto de datos para cada consulta. Por eso surge otra clase de *clustering*, llamado *grid-based*. *STING* es un tipo de este estilo; otro es *CLIQUE*, que se considera tanto basado en grillas como en densidad.

## El Algoritmo *K-Means*

### En pocas palabras

Cuando tu conjunto de datos es de tamaño pequeño a mediano y quieres encontrar el número de *clusters* que contiene, usa este algoritmo. El algoritmo encuentra el mejor valor para los centroides y agrupa todos los objetos cercanos en los *clusters* correspondientes. La optimización es NP-hard y, por lo tanto, usualmente solo proporciona soluciones aproximadas. El algoritmo realiza *hard clustering*, lo que significa que cada punto de datos pertenece únicamente a un *cluster*. Al agrupar, el algoritmo determina el mejor valor para el centroide mediante un proceso iterativo. Así funciona. Para empezar, necesitas decidir el número de *clusters* que quieres crear en tu conjunto de datos. Este es el valor $K$ en el algoritmo. Esto significa que debes estimar el número de *clusters* en el conjunto de datos antes de ejecutar el algoritmo. Esto puede ser una tarea difícil. Afortunadamente, tenemos algunas técnicas para estimar este valor de $K$. Te las explicaré en breve.  

Supongamos que decides que $K$ es igual a 3. Ahora, selecciona tres centroides al azar. Esta es la inicialización de tus *clusters*. Mide la distancia euclidiana entre cada punto de datos y el centroide. Asigna cada punto al centroide más cercano para formar los *clusters*. Calcula la media de cada *cluster*. Este será tu nuevo centroide para probar.

Repite todo el procedimiento con los nuevos centroides para crear un nuevo conjunto de *clusters*. Haz esto repetidamente hasta la convergencia, es decir, hasta que no haya más cambios o se alcance el número máximo de iteraciones que hayas decidido.  

¿El algoritmo se detiene aquí? No, aún necesitas considerar la varianza en cada *cluster*. Tu objetivo debe ser elegir centroides que minimicen la suma de los cuadrados dentro de cada *cluster*. Entonces, comienzas de nuevo con un nuevo conjunto de tres

## Algoritmo *K-Means*

Estos son los pasos del algoritmo *K-means*:  

1. Asigna un valor estimado (número de *clusters*) a $K$.  
2. Selecciona aleatoriamente $K$ puntos de datos como centroides.  
3. Repite los siguientes pasos hasta la convergencia (no se observan más cambios) o hasta alcanzar el número de iteraciones predefinido:  
   (a) Calcula la suma de las distancias al cuadrado entre todos los puntos de datos y los centroides.  
   (b) Asigna cada punto de datos al centroide más cercano.  
   (c) Calcula los nuevos centroides tomando la media de todos los puntos de datos pertenecientes a cada *cluster*. 

### Función Objetivo

Dado un conjunto de $n$ observaciones $(x_1, x_2, ..., x_n)$, donde cada observación es un vector real de $d$ dimensiones, nuestro objetivo es dividir todas las observaciones en $k$ ($k \le n$) conjuntos $\{S_1, S_2, ..., S_k\}$ de manera que se minimice la WCSS (suma de cuadrados dentro del grupo) (*within-cluster sum of squares*).

Matemáticamente, el objetivo se expresa como:  
$$
\arg\min_S \sum_{i=1}^{k} \sum_{x \in S_i} \|x - \mu_i\|^2
$$
donde $\mu_i$ es la media de los puntos en $S_i$. 


Este es un problema NP-hard, donde la complejidad temporal es $O(n^{dk+1})$, siendo $n$ el número de observaciones, $k$ el número de *clusters* y $d$ la dimensión. Ten en cuenta que necesitarás repetir este proceso de encontrar *clusters* múltiples veces y luego seleccionar el mejor conjunto de *clusters*.

### Selección de *Clusters* Óptimos

Cuando observas el conjunto de datos que deseas dividir en *clusters*, es difícil visualizar cuántos *clusters* podrían formarse. Afortunadamente, tenemos algunas técnicas disponibles para este propósito. Las llamamos:  

• *Elbow*  
• *Silhouette*  
• *Gap statistics*  

#### Método del Codo (*Elbow*)

Este es probablemente el método más popular, basado en el concepto de WCSS (*within-cluster sum of squares*). Como se vio anteriormente, WCSS se expresa matemáticamente como:  
$$
WCSS = \sum_{i=1}^{k} \sum_{x \in S_i} \|x - \mu_i\|^2
$$
donde la distancia puede ser la distancia euclidiana, como típicamente se usa en el algoritmo de *k-nearest neighbors* (clasificación). 

Ejecutamos el algoritmo para diferentes valores de $K$, por ejemplo, en el rango de 1 a 10, midiendo WCSS para cada uno. Luego, graficamos WCSS vs $K$, como se ve en la figura 9.4.

#### Método de Silueta (Silhouette)

Este método se basa en medir la calidad del *cluster* determinando qué tan bien se encuentra cada objeto dentro de su *cluster*. Para medir la calidad, definimos un coeficiente de *silhouette*:  
$$
\text{Coeficiente de Silhouette } (s_i) = \frac{y - x}{\max(x, y)}
$$
donde $y$ es la distancia intracluster promedio, es decir, la distancia media a los otros puntos de datos en el mismo *cluster*. $x$ representa la distancia promedio al *cluster* más cercano, que es la distancia media a otros puntos de datos del *cluster* más próximo. El valor de $s_i$ varía entre -1 y 1. Un valor cercano a 1 indica que el punto de datos está lo suficientemente cerca de su *cluster* para considerarse parte de él. Un valor cercano a -1 indica que hemos asignado el punto de datos al *cluster* equivocado.  

Con esta definición del coeficiente de *silhouette*, diseñamos el algoritmo de manera similar al método *elbow*. Ejecutamos el algoritmo para diferentes valores de $K$, por ejemplo, en el rango de 1 a 10. Para cada $K$, calculamos el *silhouette* promedio y luego graficamos este puntaje contra los valores de $K$. El pico de la curva, es decir, el puntaje máximo de *silhouette*, determina el mejor valor de $K$. La figura 9.5 muestra dicho gráfico.

#### Método de Estadística de Brecha (Gap Statistics)

Mientras que el método de silueta mide la calidad, la estadística de de brecha (*gap*) mide qué tan bueno es el *clustering*. Investigadores de Stanford desarrollaron este método. Compara el $\log W_k$ (*within-cluster sum of squares*) con una distribución de referencia nula de los datos.  

Una distribución nula es una distribución sin un *clustering* evidente. El valor óptimo de $K$ es aquel para el cual este $\log$ cae más lejos por debajo de la curva de referencia. Matemáticamente, el *gap* se expresa como:  
$$
\text{Gap}_n(k) = E^*_n\{\log W_k\} - \log W_k
$$

Generas el conjunto de datos de referencia muestreando uniformemente dentro del rectángulo delimitador del conjunto de datos original. En la figura 9.6 se observa que el *gap* es máximo para $K = 3$. Este será el valor óptimo para el *clustering*.

### Implementación

La librería *sklearn* proporciona la implementación del algoritmo *K-means* en la clase *sklearn.cluster.KMeans*. Usar esta clase es bastante sencillo: solo necesitas especificar el valor de $K$ durante su instanciación; el resto de los parámetros puede tomar sus valores por defecto. La clase también soporta *K-means++*, donde puedes especificar tus centroides deseados al inicio.  
Para ilustrar la implementación, he creado un pequeño proyecto que también muestra los tres métodos para obtener el $K$ óptimo.

### Proyecto

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns;
from sklearn.metrics import pairwise_distances
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

sns.color_palette("Set2")
%matplotlib inline
```

Primero, creamos un conjunto de datos de 500 muestras aleatorias usando el método *make_blobs*.

```python
X, Y = make_blobs(n_samples=500, random_state=0, centers=3, n_features=2, cluster_std=0.9)
```

Los datos generados se ven como los mostrados en la figura 9.8.  

```python
plt.rcParams["figure.figsize"] = (6,6)   
plt.grid()
plt.scatter(X[:, 0], X[:, 1], s=50, c='blue');
```

<p style="color:blue; margin-top: 5px; font-weight: bold;">Fig. 9.8</p>

<p>Conjunto de datos aleatorio.</p>

Ahora agrupamos el conjunto de datos llamando a la función *KMeans* proporcionada por la librería.

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

Después del *clustering*, haremos la predicción sobre el mismo conjunto de datos. La figura 9.9  muestra los resultados de agrupamiento (*clustering*).

```python
km_pred = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=km_pred, cmap='viridis')
centers = kmeans.cluster_centers_
plt.rcParams["figure.figsize"] = (6,6)
plt.grid()
# display centroid in black
plt.scatter(centers[:, 0], centers[:, 1], c='black');
```

<p style="color:blue; margin-top: 5px; font-weight: bold;">Fig. 9.9</p>

<p>Conjunto de datos agrupado.</p>

Puedes ver lo fácil que es agrupar un conjunto de datos. El único requisito es una buena estimación del número de *clusters*. En el proyecto, también he proporcionado el código de implementación para los métodos *elbow*, coeficiente de *silhouette* y estadística de *gap*.

### Representación Gráfica del Método del Codo (*Elbow*)

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

elbow = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=101)
    kmeans.fit(X)
    elbow.append(kmeans.inertia_)

plt.figure(figsize=(6, 4))
sns.lineplot(x=range(1, 10), y=elbow, color='blue')
plt.title('Método del Codo')
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.grid()
plt.show()
```

### Representación Gráfica del Método del Silueta (*Silhouette*)

```python
from sklearn.metrics import silhouette_score

k_clusters = []
sil_coeffecients = []

for n_cluster in range(2,11):
    kmeans = KMeans(n_clusters = n_cluster).fit(X)
    label = kmeans.labels_
    sil_coeff = silhouette_score(X, label)
    print("For n_clusters={}, Silhouette Coefficient = {}".format(n_cluster, sil_coeff))
    sil_coeffecients.append(sil_coeff)
    k_clusters.append(n_cluster)

plt.rcParams["figure.figsize"] = (6, 4)    
plt.plot(k_clusters, sil_coeffecients)
plt.ylabel('Coeficinte de Silueta'), plt.xlabel('Número de Clusters')
plt.title ('Método de Silueta')
plt.show()


sil_best = max(sil_coeffecients)
k_best_index = sil_coeffecients.index(sil_best)
k_best = k_clusters[k_best_index]
print("Optimum Number of Clusters:", k_best)
```

### Cálculo y Representación de la Estadística de Brecha

```python
def compute_inertia(a, X):
    W = [np.mean(pairwise_distances(X[a == c, :])) for c in np.unique(a)]
    return np.mean(W)

def compute_gap(clustering, data, k_max=5, n_references=5):
    if len(data.shape) == 1:
        data = data.reshape(-1, 1)
    reference = np.random.rand(*data.shape)
    reference_inertia = []
    for k in range(1, k_max+1):
        local_inertia = []
        for _ in range(n_references):
            clustering.n_clusters = k
            assignments = clustering.fit_predict(reference)
            local_inertia.append(compute_inertia(assignments, reference))
        reference_inertia.append(np.mean(local_inertia))
    
    ondata_inertia = []
    for k in range(1, k_max+1):
        clustering.n_clusters = k
        assignments = clustering.fit_predict(data)
        ondata_inertia.append(compute_inertia(assignments, data))
        
    gap = np.log(reference_inertia)-np.log(ondata_inertia)
    return gap, np.log(reference_inertia), np.log(ondata_inertia)

k_max = 5
gap, reference_inertia, ondata_inertia = compute_gap(KMeans(), X, k_max)


plt.plot(range(1, k_max+1), reference_inertia,
         '-o', label='reference')
plt.plot(range(1, k_max+1), ondata_inertia,
         '-o', label='data')
plt.xlabel('k')
plt.ylabel('log(inercia)')
plt.title ('Gráfico de inercia')
plt.show()
```

## El Algoritmo *K-Medoids*
Al igual que *K-Means*, *K-Medoids* es un algoritmo de *clustering* no supervisado que supera algunas limitaciones de *K-means*.

### En pocas palabras

En el *clustering K-means*, has visto que seleccionamos el promedio de los puntos de un *cluster* como su centro. Este punto promedio puede no ser un punto de datos real en tu conjunto de datos. En *K-medoids*, siempre elegimos un punto de datos real como el centro del *cluster*. Este centro se llama *medoid* o *exemplar*. Es el punto de datos más central del *cluster*, con la suma mínima de distancias a los otros puntos. Hay otro algoritmo de *clustering* en esta categoría, que es *K-medians*, en el cual la mediana se convierte en el centro del *cluster*.

## ¿Por qué K-Medoids?

El algoritmo *K-means* es sensible a los valores atípicos, mientras que el algoritmo *K-medoids* es más robusto frente a ruidos y valores atípicos. Para demostrar esto, creé un conjunto de datos aleatorio, calculé *K-means* y *K-medoids* sobre él; luego agregué algunos valores atípicos al conjunto de datos y recalculé los dos valores. El desplazamiento del centroide en *K-means* fue mucho mayor que el observado en *K-Medoids*. La figura 9.10 muestra estas observaciones.

La razón detrás de esta robustez frente al ruido y los valores atípicos es que *K-medoids* minimiza una suma de disimilitudes por pares, en lugar de una suma de distancias euclidianas al cuadrado, como ocurre con *K-means*. Tanto *K-medoids* como *K-means* son algoritmos de partición que dividen el conjunto de datos en grupos para minimizar la distancia entre el centro del clúster y los otros puntos asignados a ese clúster.

Otra razón para elegir *K-medoids* sobre *K-means* es que podemos usar este algoritmo con medidas de disimilitud arbitrarias, a diferencia de *K-means*, donde usamos distancia euclidiana para soluciones eficientes. En *K-medoids*, puedes usar Manhattan, Euclidiana o incluso coseno. Te mostraré el *clustering* hecho con estas tres medidas de disimilitud diferentes en el proyecto que sigue.
Como existen muchas soluciones heurísticas, *K-medoids* es un problema NP-difícil. Veamos ahora el algoritmo en sí.

### Algoritmo

El algoritmo K-medoids, también conocido como *PAM* (*partitioning around medoids*) usa una búsqueda voraz para encontrar la solución óptima. La búsqueda voraz funciona más rápido que una búsqueda exhaustiva. El algoritmo requiere que introduzcas el valor de $K$, el cual estimamos usando métodos como silueta. Escribimos el algoritmo de la siguiente manera:

1. Inicializar: seleccionar de manera voraz $K$ de entre los $n$ puntos de datos como los *medoids*.  
2. Asociar cada punto de datos al *medoid* más cercano usando cualquier métrica de distancia arbitraria.  
3. Realizar el siguiente intercambio, mientras el costo de la configuración disminuya:  
   (a) Sea $m$ el *medoid* y $o$ el punto de datos que no es *medoid*.  
   (i) Intercambiar $m$ y $o$, calcular el cambio de costo.  
   (ii) Si el cambio de costo es el mejor actual, recordar este $m$ y $o$ como una combinación.  
4. Realizar el intercambio de la mejor combinación de $m$ y $o$, si disminuye el costo. De lo contrario, terminar el algoritmo.  

### Méritos y Deméritos

El algoritmo *PAM* proporciona flexibilidad en la selección de métodos de medida de similitud. También es robusto al ruido y a los valores atípicos en comparación con K-means. La desventaja es que no podemos aplicarlo a conjuntos de datos grandes, ya que es computacionalmente costoso.
