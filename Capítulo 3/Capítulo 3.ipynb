## En Pocas Palabras

La regresión es uno de los algoritmos mejor comprendidos y estudiados en estadística. Ha sido aplicada con éxito al aprendizaje automático y se ha utilizado ampliamente en muchas aplicaciones prácticas. Es una técnica de modelado predictivo que establece la relación entre una variable dependiente y una o más variables independientes. Existen muchas formas de regresión; la regresión lineal y la regresión logística son las más utilizadas. ## Regresión Lineal

Probablemente sea el primer tipo de análisis de regresión estudiado rigurosamente y utilizado de forma extensiva. La regresión lineal asume una relación lineal entre el objetivo y sus predictores. Esto facilita el ajuste del modelo en tales conjuntos de datos, y los estimadores resultantes pueden validarse con facilidad.

En la regresión lineal, se busca la recta de mejor ajuste que establezca una relación entre una variable dependiente *(y)* y una o más variables independientes *(x)*. Las variables independientes pueden ser continuas o discretas. Es necesario encontrar la línea de mejor ajuste, también llamada línea de regresión. Para cualquier valor de x no observado, se puede determinar el valor correspondiente de y resolviendo matemáticamente la ecuación de la línea de regresión.
Matemáticamente, la ecuación de la línea de regresión se expresa como:

$$
y_i = \beta_0 + \beta_1 x_{i1} + \cdots + \beta_p x_{ip} + \varepsilon_i = x_i^T \beta + \varepsilon_i, \quad i = 1, \dots, n
$$

donde el superíndice $T$ denota la transpuesta, de modo que $x_i^T \beta$ es el producto interno entre los vectores $x_i$ y $\beta$.  
$\varepsilon$ es el término de ruido en la relación lineal entre la variable dependiente y los regresores.



Generalmente, estas n ecuaciones se escriben en notación matricial como:  

$$
y = X \beta + \varepsilon
$$


donde:  

$$
y =
\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_n
\end{bmatrix}
$$



$$
X =  
\begin{bmatrix}
x_1^T \\
x_2^T \\
\vdots \\
x_n^T
\end{bmatrix} =
\begin{bmatrix}
1 & x_{11} & \cdots & x_{1p} \\
1 & x_{21} & \cdots & x_{2p} \\
\vdots & \vdots & \ddots & \vdots \\
1 & x_{n1} & \cdots & x_{np}
\end{bmatrix}
$$

$$
\beta =
\begin{bmatrix}
\beta_0 \\
\beta_1 \\
\beta_2 \\
\vdots \\
\beta_p
\end{bmatrix}
$$

$$
\varepsilon =
\begin{bmatrix}
\epsilon_1 \\
\epsilon_2 \\
\vdots \\
\epsilon_n
\end{bmatrix}
$$

Se utiliza el método de mínimos cuadrados para obtener la línea de mejor ajuste. En este método, se minimiza la suma de los cuadrados de las desviaciones verticales de cada punto de datos observado con respecto a la línea de regresión. Matemáticamente, se expresa como:

$$
\min_{\beta} \| X\beta - y \|_2^2
$$

### Suposiciones

En la técnica de estimación de mínimos cuadrados, se realizan ciertas suposiciones sobre las variables predictoras, la variable objetivo y su relación. Los estadísticos han ampliado esta técnica de estimación estándar con muchas extensiones que se discutirán más adelante. Aunque estas extensiones hacen que el proceso de estimación sea más complejo y consuma más tiempo, reducen las suposiciones a una forma más débil e, incluso, a veces las eliminan por completo. Veamos las suposiciones.

Se asume que no existe error en la medición de las variables predictoras x, lo cual es una suposición poco realista. Si hay un error en la medición de las variables predictoras, se supone que la varianza de los errores es independiente de sus valores. A esto se le llama **homocedasticidad**, es decir, la suposición de varianza constante. Generalmente, esto no es cierto: la desviación estándar de los valores predictivos es proporcional a su media, en lugar de ser constante.  

Otra suposición es que los errores en las variables de respuesta son no correlacionados entre sí. La multicolinealidad en las variables predictoras puede afectar seriamente la estimación de los parámetros. Lo más importante es la suposición de **linealidad**, según la cual la media de la variable de respuesta es una función lineal de los coeficientes de regresión y de las variables predictoras.

A partir de estas suposiciones, lo que se quiere destacar es que no se puede aplicar la regresión lineal directamente a cualquier conjunto de datos para análisis predictivo. Se requiere un análisis exhaustivo del conjunto de datos (EDA), y pueden ser necesarias algunas transformaciones o escalado de los puntos de datos antes de aplicar la regresión lineal. Después de todo, la relación podría ser polinómica o corresponder a algún hiperplano complejo. Durante el análisis exploratorio, si se observan tales tendencias, se deben aplicar otras técnicas de regresión que se discuten más adelante.

## Regresión Polinómica

En este tipo de regresión, modelamos la relación entre la variable independiente x y la variable dependiente y como un polinomio de grado n en x. Matemáticamente, se expresa como:

$$
y_i = \beta_0 + \beta_1 x_i + \beta_2 x_i^2 + \dots + \beta_m x_i^m + \epsilon_i, \quad i = 1, 2, \dots, n
$$


En notación matricial, el modelo completo se escribe como:

$$
\vec{y} = \mathbf{X} \vec{\beta} + \vec{\epsilon}
$$

A partir de estas ecuaciones, se puede entender que la regresión polinómica es, técnicamente, un caso especial de la regresión lineal múltiple. La mejor aproximación se encuentra tratando cada \(x_i\) como una variable independiente y aplicando la regresión lineal a cada una. La combinación de estas regresiones múltiples da como resultado un polinomio final.

## Regresión Ridge

Cuando se observa multicolinealidad (las variables independientes están altamente correlacionadas) en un conjunto de datos, se utiliza la regresión Ridge. Ridge es un caso especial de la regularización de Tikhonov. Regulariza todos los parámetros por igual. Generalmente, la multicolinealidad se observa en conjuntos de datos con muchas características. Esta técnica mejora las estimaciones de los parámetros a costa de un sesgo tolerable, agregando cierto grado de sesgo a las estimaciones de regresión para reducir los errores estándar.


Matemáticamente, se expresa como:

$$
\arg\min_{\beta \in \mathbb{R}^p} \underbrace{\| y - X\beta \|_2^2}_{\text{Loss}} + \underbrace{\lambda \|\beta\|_2^2}_{\text{Penalty}}
$$

donde el primer término es nuestro término de mínimos cuadrados y el segundo es el multiplicador de Lagrange de la restricción. Cuando \(\lambda = 0\), el estimador se reduce a los mínimos cuadrados ordinarios. Para cualquier otro valor, se contrae el término de mínimos cuadrados para lograr una baja varianza. Esto corresponde a la regularización L2.

## Regresión Lasso

Al igual que en Ridge, Lasso también penaliza el tamaño absoluto de los coeficientes de regresión. Matemáticamente, se expresa como:

$$
\arg\min_{\beta \in \mathbb{R}^p} \underbrace{\| y - X\beta \|_2^2}_{\text{Loss}} + \underbrace{\lambda \|\beta\|_1}_{\text{Penalty}}
$$

El término de penalización ahora es el valor absoluto en lugar de los cuadrados. Una penalización grande reduce la estimación hacia cero absoluto, lo que puede resultar en la eliminación de la variable. Esto ayuda en la selección de características. Lasso es un método de regularización L1. Cuando se tiene un grupo de predictores altamente correlacionados, Lasso selecciona uno de ellos y reduce los demás a cero.

## Regresión ElasticNet

Esta es una combinación de las regresiones Lasso y Ridge. Matemáticamente se expresa como:

$$
\hat{\beta} = \arg\min_{\beta} ( \| y - X\beta \|_2^2 + \lambda_2 \|\beta\|_2^2 + \lambda_1 \|\beta\|_1 )
$$


Como se puede ver, el modelo se entrena utilizando regularizadores *l1* y *l2*. Si el conjunto de datos tiene múltiples predictores correlacionados, ElasticNet probablemente seleccionará todos, mientras que Lasso escogerá uno al azar. De esta forma, se fomenta un efecto de grupo sobre predictores altamente correlacionados, sin limitación en el número de predictores seleccionados. La desventaja es que puede sufrir de doble contracción.

### Modelado del Conjunto


Se construirán y compararán los siguientes modelos utilizando sus correspondientes métricas de error:

- Regresión Lineal.
- Regresión Ridge con validación cruzada.
- Regresión Lasso con validación cruzada.

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

def error_metrics(y_pred,y_val):
  print('MSE: ',mean_squared_error(y_pred,y_val))
  print('RMSE: ',np.sqrt(mean_squared_error(y_pred,y_val)))
  print('Coefficient of determination: ',r2_score(y_pred,y_val))
```

#### Métricas de error


**Error Cuadrático Medio (MSE)**: Calcula el cuadrado de las diferencias entre los valores predichos y los valores reales por instancia, luego los suma y finalmente obtiene el promedio.

**Raíz del Error Cuadrático Medio (RMSE)**: Consiste en calcular la raíz cuadrada del MSE.

**Coeficiente de Determinación (R²)**: Toma en cuenta el cuadrado de las diferencias y el cuadrado de la diferencia de los valores reales con la media de los valores. Como se puede ver en la fórmula, a medida que el numerador disminuye, la métrica tiende a 1.0, lo que indica la proporción de la varianza en la variable dependiente (etiqueta) que es predecible a partir de las variables independientes.
Idealmente, esta métrica debería estar muy cerca de 1.0 para contar con un modelo robusto.

```python
from sklearn.model_selection import train_test_split

X_train, X_val, label_train, label_val = train_test_split(df, label, test_size=0.15, random_state=42)
     
X_train.shape, label_train.shape, X_val.shape, label_val.shape
```

### Regresión Ridge

Sklearn implementa la regresión Ridge con validación cruzada incorporada en la clase `sklearn.linear_model.RidgeCV`. ## Regresión Lineal Bayesiana

Hasta ahora, has aprendido sobre la regresión lineal simple y los efectos de las regularizaciones. ¿Funcionarán bien estos modelos si cuentas con datos insuficientes o mal distribuidos? Aquí es donde la regresión lineal bayesiana (BLR, por sus siglas en inglés) te resulta útil. BLR no solo ayuda a encontrar el valor óptimo único para los parámetros del modelo, sino que también permite determinar la distribución posterior de dichos parámetros.  

La regresión es una tarea de aprendizaje automático usada para predecir valores continuos, a diferencia de la clasificación, que se utiliza para predecir valores discretos. Cuando escuchamos la palabra "Bayesiano", solemos pensar en Naive Bayes. Sin embargo, los principios bayesianos también pueden aplicarse para realizar regresión.


En primer lugar, debemos entender que lo bayesiano es simplemente un enfoque para definir y estimar modelos estadísticos. La regresión bayesiana puede ser muy útil cuando tenemos pocos datos en el conjunto o cuando estos están mal distribuidos. Los métodos de regresión bayesiana son muy poderosos, ya que no solo nos proporcionan estimaciones puntuales de los parámetros de la regresión, sino que entregan una distribución completa sobre dichos parámetros. Esto permite asignar un prior tanto a los coeficientes como al ruido, de modo que en ausencia de datos, los priors puedan prevalecer. Más importante aún, la regresión lineal bayesiana nos permite preguntar en qué partes de su ajuste a los datos el modelo es realmente confiado y en cuáles la incertidumbre es alta (quizá dependiendo casi por completo de los priors).

La BLR es el enfoque bayesiano al análisis de regresión lineal. Una medida de ambigüedad podría ser la desviación estándar de las predicciones de todos los modelos, algo que los estimadores puntuales no proporcionan por defecto. Saber lo que el modelo no sabe ayuda a hacer la inteligencia artificial más comprensible. En el análisis de regresión, nos fijamos en la distribución condicional de la variable de respuesta en distintos niveles de una variable predictora.

BLR se basa en el famoso teorema de Bayes. La estadística bayesiana es un método matemático basado en probabilidades para resolver problemas estadísticos, que ayuda a formar creencias a partir de la evidencia de nuevos datos. Por ello, esta técnica es ideal cuando no se dispone de suficientes puntos de datos en el conjunto de datos.

$$
P(c \mid x) = \frac{P(x \mid c) \cdot P(c)}{P(x)}
$$

<p style="color:blue; margin-top: 5px; font-weight: bold;">Fig. 3.5</p>

<p>Probabilidad posterior en el teorema de Bayes.</p>

donde:  
- $P(c \mid x)$ es la probabilidad posterior de la clase $c$ dado el predictor $x$  
- $P(c)$ es la probabilidad a priori de la clase $c$  
- $P(x \mid c)$ es la probabilidad (verosimilitud) del predictor dado la clase $c$  
- $P(x)$ es la probabilidad a priori del predictor $x$

Aplicando este teorema a tu conjunto de datos, te permite establecer una distribución a priori tanto sobre los coeficientes como sobre el ruido; utilizas estas distribuciones a priori en ausencia de datos. Más importante aún, la regresión lineal bayesiana (BLR) te informa sobre qué partes de los datos tiene confianza y cuáles se basan completamente en las distribuciones a priori. Específicamente, puedes investigar lo siguiente:

- ¿Cuál es su confianza en la relación estimada?

- ¿Cuál es el ruido estimado y su distribución posterior?

- ¿Cuál es el gradiente estimado y su distribución posterior?

Sin entrar en las matemáticas de la BLR, veamos su implementación.

#### Ventajas de la Regresión Lineal Bayesiana  

- Gracias al procesamiento bayesiano, se recupera toda la gama de soluciones inferenciales, en lugar de una estimación puntual y un intervalo de confianza como en la regresión clásica.  

- Se obtiene un enfoque totalmente fundamentado matemáticamente, donde es posible incorporar fácilmente el conocimiento previo sobre el dominio del problema.  

- Funciona bien cuando se dispone de pocos datos y ofrece más que una simple predicción.  

#### Desventajas de la Regresión Lineal Bayesiana  

- No es adecuada cuando no se trata de un problema de regresión, cuando un modelo lineal no funciona bien, o si no se desea utilizar un enfoque bayesiano.  

- Encontrar un buen prior puede ser difícil y, cuando se dispone de muchos datos y solo interesa aprender un único modelo en lugar de una distribución sobre modelos, se termina obteniendo esencialmente el mismo modelo que con otros tipos de regresión lineal.

```python
import warnings # To hide the warnings
import itertools # Library for efficient looping
import numpy as np # Scientific Calculations
import pandas as pd # Data Structures
import matplotlib.pyplot as plt # Visualizations
import seaborn as se # Visualizations
from sklearn.linear_model import BayesianRidge
from sklearn.model_selection import train_test_split,cross_val_score # Train and Test Split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error # Scoring Metrics for Regression
warnings.filterwarnings('ignore')
```

### Proyecto BLR

El proyecto utiliza un conjunto de datos sobre una planta de energía de ciclo combinado publicado en el repositorio de UCI. El conjunto de datos contiene aproximadamente 9000 puntos de datos.  

La salida de la planta (**PE**) depende de cuatro características: Temperatura (**AT**), Vacío de escape (**V**), Presión ambiental (**AP**) yHumedad relativa (**RH**) con el objetivo es predecir la energía eléctrica neta por hora (**EP**) de la planta.  

Una planta de energía de ciclo combinado (**CCPP**) está compuesta por: Turbinas de gas (**GT**), Turbinas de vapor (**ST**) y Generadores de vapor de recuperación de calor.

En una CCPP, la electricidad se genera mediante turbinas de gas y de vapor, las cuales están integradas en un solo ciclo, transfiriéndose la energía de una turbina a otra. Mientras que el vacío se recoge de la turbina de vapor y la afecta directamente, las otras tres variables ambientales influyen en el rendimiento de la turbina de gas.  

Después de leer el conjunto de datos, se configurarán las características y el objetivo utilizando las siguientes sentencias.  

#### Información de los Atributos:

Las características consisten en variables ambientales promedio por hora:  

- Temperatura (T) en el rango 1.81°C a 37.11°C  
- Presión Ambiental (AP) en el rango 992.89 a 1033.30 milibar  
- Humedad Relativa (RH) en el rango 25.56% a 100.16%  
- Vacío de Escape (V) en el rango 25.36 a 81.56 cm Hg  
- Energía eléctrica neta por hora (EP) en el rango 420.26 a 495.76 MW  

Los promedios se toman de varios sensores ubicados en la planta que registran las variables ambientales cada segundo. Las variables se proporcionan sin normalización.  


```python
df=pd.read_csv('power.csv')
df.head()
```

Después de leer el conjunto de datos, configur las funciones y el objetivo utilizando declaraciones:

```python
features=['AT','V','AP','RH']

target='PE'
```

Establece algunos valores de parámetros:

#### Parámetros de Ajuste del Modelo

- alpha_1: parámetro de forma para la distribución Gamma como prior sobre el parámetro alpha.
- alpha_2: parámetro de escala inversa (parámetro de tasa) para la distribución Gamma como prior sobre el parámetro alpha.
- lambda_1: parámetro de forma para la distribución Gamma como prior sobre el parámetro lambda.
- lambda_2: parámetro de escala inversa (parámetro de tasa) para la distribución Gamma como prior sobre el parámetro lambda.

```python
Alpha1=0.00006138969568741323
Alpha2=12267443.02717635
Lambda1=5.936316029785909e-10
Lambda2=3075473.158359934
```

#### Selección de Características

Es el proceso de reducir el número de variables de entrada al desarrollar un modelo predictivo. Se utiliza para disminuir tanto el costo computacional del modelado como, en algunos casos, para mejorar el rendimiento del modelo.
Asignaremos todas las características de entrada requeridas a X y la variable objetivo/resultante a Y.

```python
X=df[features]
Y=df[target]
```

#### División de Datos


La división en entrenamiento y prueba es un procedimiento para evaluar el rendimiento de un algoritmo. Este procedimiento consiste en tomar un conjunto de datos y dividirlo en dos subconjuntos. El primer subconjunto se utiliza para ajustar/entrenar el modelo. El segundo subconjunto se utiliza para realizar predicciones. El objetivo principal es estimar el rendimiento del modelo con datos nuevos.

```python

split_ratio=0.20
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=split_ratio,random_state=123)
```

#### Modelo

La regresión bayesiana puede ser muy útil cuando disponemos de pocos datos en el conjunto o cuando los datos están mal distribuidos. El resultado de un modelo de regresión bayesiana se obtiene a partir de una distribución de probabilidad, en comparación con las técnicas de regresión convencionales donde el resultado se obtiene simplemente a partir de un valor único de cada atributo. Las técnicas de regresión bayesiana pueden emplearse para incluir parámetros de regularización en el procedimiento de estimación: el parámetro de regularización no se establece de forma rígida, sino que se ajusta a los datos disponibles.

Aplicaremos BLR en el conjunto de datos de entrenamiento de la siguiente manera:

```python
model=BayesianRidge(alpha_1=Alpha1,alpha_2=Alpha2,lambda_1=Lambda1,lambda_2=Lambda2)
model.fit(x_train,y_train)
```

Después del entrenamiento, hay que verificar la precisión del modelo en los datos de prueba:

```python
y_pred=model.predict(x_test)
print("Accuracy score {:.2f} %\n".format(model.score(x_test,y_test)*100))
```

También se puede imprimir el R2, MAE y MSE de la siguiente manera:

```python
print("Puntaje R2 (R2 Score): {:.2f} %".format(
r2_score(y_test,y_pred)*100))
print("Error Absoluto Medio (MAE): {:.2f}".format(
mean_absolute_error(y_test,y_pred)))
print("Mean Squared Error (MSE): {:.2f}".format(
mean_squared_error(y_test,y_pred)))
```

Puntaje R2: La función _r2_score_ calcula el porcentaje de variabilidad explicado por nuestro modelo, ya sea como fracción o como proporción de predicciones correctas.

MAE: La función de Error Absoluto Medio calcula la cantidad de error total (distancia absoluta promedio entre los datos reales y los datos predichos) de nuestro modelo.

MSE: La función de Error Cuadrático Medio eleva al cuadrado el error (penaliza al modelo por errores grandes) de nuestro modelo.


#### Gráfico de Predicciones

Primero, se utiliza un gráfico para mostrar las observaciones reales, con `x_train` en el eje x y `y_train` en el eje y. Para la línea de regresión, se emplea `x_train` en el eje x y las predicciones de las observaciones de `x_train` en el eje y.

## Regresión Logística

A diferencia de la regresión lineal, donde las predicciones son continuas, en la regresión logística estas son discretas. El resultado será un valor binario, como aprobar/reprobar, verdadero/falso, sí/no, etc. Cuando la salida es binaria, se llama regresión logística binaria.

Los valores de la variable dependiente pueden ser ordinales, como bajo, medio y alto; en este caso se llama regresión logística ordinal. Si la variable dependiente es multiclase, como gato, perro, león y cabra, se llama regresión logística multinomial. En todos los casos, la salida es discreta, por lo que la regresión logística se considera un algoritmo de clasificación.

Algunos ejemplos prácticos de regresión logística incluyen la clasificación de correos electrónicos (spam o no spam), la detección de tumores (maligno o benigno) y la verificación de transacciones en línea (fraude o no fraude).  

Básicamente, la regresión logística se logra pasando la salida de la regresión lineal a través de una función de activación que mapea un número real a 0 o 1. Esto se ilustra en la figura 3.7.

La función sigmoide, mostrada en la figura 3.8, se usa como función de activación. Su valor está entre 0 y 1. Si la probabilidad es mayor que 0.5, se considera que la salida es 1; de lo contrario, 0.

Matemáticamente, la ecuación de la regresión logística se expresa como:

$$
P = \frac{1}{1 + e^{-(a + bX)}}
$$

donde \(P\) es la estimación de probabilidad.

#### Ingeniería de Características para el Proyecto

Tenemos 2 tipos de características categóricas: binarias y nominales. Como sabemos, las binarias pueden tomar valores 1 o 0, mientras que las nominales deben ser codificadas con one-hot usando `get_dummies`.

```python
df_2.head()
```

Binarización de las características de Gender y EverBenched:

```python
df_2['Gender']=df_2['Gender'].replace('Male',1).replace('Female',0)

df_2['EverBenched']=df_2['EverBenched'].replace('Yes',1).replace('No',0)
```

Codificación única de las características de Ciudad (City) y Educación (Education):

```python
cat_features=['City', 'Education']
df_2 = pd.get_dummies(df_2, columns=cat_features, drop_first=True)
```

Arriba especifiqué `drop_first=True` para omitir una de las clases, porque esa clase es equivalente cuando todas las demás clases son 0. Este método, aplicado a conjuntos de datos con cientos de características categóricas, ayuda a reducir la maldición de la dimensionalidad.

```python
df_2
```

```python
label=df_2['LeaveOrNot']

df_2.drop(['LeaveOrNot'], axis=1, inplace=True)

df_2
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, label_train, label_val = train_test_split(
    df_2, label,
    test_size=0.20,
    random_state=42
    )

X_train.shape, label_train.shape, X_val.shape, label_val.shape
```

```python
from sklearn.preprocessing import StandardScaler
s = StandardScaler()

X_train_scaled = s.fit_transform(X_train)
X_val_scaled = s.transform(X_val)
```

### Modeling

Como antes, se utilizarán algunos fragmentos de código aquí para mostrar cómo usar la clase de regresión logística. Se aplica la regresión logística sobre el conjunto de datos usando el siguiente fragmento de código:

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver='liblinear').fit(X_train_scaled, label_train)
```

Aplica la regularización _l1_ de la siguiente manera:

```python
from sklearn.linear_model import LogisticRegressionCV

lr_l1 = LogisticRegressionCV(Cs=15, cv=4, penalty='l1', solver='liblinear').fit(
        X_train_scaled, label_train)  #C=10
```

Y la regularización l2 como en la siguiente línea:

```python
lr_l2 = LogisticRegressionCV(Cs=15, cv=4, penalty='l2', solver='liblinear').fit(
    X_train_scaled, label_train)  #C=10
```

Se predice la clase y la probabilidad para cada una:

```python
y_pred = list()
y_prob = list()

coeff_labels = ['lr', 'l1', 'l2']
coeff_models = [lr, lr_l1, lr_l2]

for lab,mod in zip(coeff_labels, coeff_models):
    y_pred.append(pd.Series(mod.predict(X_val_scaled), name=lab))
    y_prob.append(pd.Series(mod.predict_proba(X_val_scaled).max(axis=1), name=lab))
    
y_pred = pd.concat(y_pred, axis=1)
y_prob = pd.concat(y_prob, axis=1)

y_pred.head()
```

```python
y_prob.head()
```

A continuación estamos mostrando todas las filas que tienen clases predichas diferentes entre los modelos (`lr != l1`). Podemos observar esto para todas las columnas de una manera más clara usando un gráfico de matriz de confusión.

```python
y_pred[y_pred['lr']!=y_pred['l1']] 
```

### Cálculo de métricas de error:  

Podemos ver en detalle las métricas de error para los 3 modelos usando el `classification_report`. Se imprime el reporte de clasificación usando la clase `classification_report` provista por la librería `sklearn.metrics`. Esto se ilustra en el siguiente fragmento de código:

El resultado del fragmento de código se muestra en la figura 3.9.

```python
from sklearn.metrics import classification_report
print('Classification report for Logistic regression without regularization:')
print(classification_report(label_val,y_pred['lr']))

print('Classification report for Logistic regression with L1(Lasso) regularization:')
print(classification_report(label_val,y_pred['l1']))
    
print('Classification report for Logistic regression with L2(Ridge) regularization:')
print(classification_report(label_val,y_pred['l2']))
```

<p style="color:blue; margin-top: 5px; font-weight: bold;">Fig. 3.9</p>

<p>Informes de clasificación variados.</p>

Para resumir y promediar los valores obtenidos haremos lo siguiente:

```python
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
from sklearn.preprocessing import label_binarize

metrics = list()
cm = dict()

for lab in coeff_labels:

    # Precisión, recuperación, puntuación f de la función de soporte multiclase, 
    # los promediaremos porque tendremos un valor por clase
    precision, recall, fscore, _ = score(label_val, y_pred[lab], average='weighted')
    
    # La forma habitual de calcular la precisión
    accuracy = accuracy_score(label_val, y_pred[lab])
    
    # Las puntuaciones ROC-AUC se pueden calcular binarizando los datos
    auc = roc_auc_score(label_binarize(label_val, classes=[0,1]),
                        label_binarize(y_pred[lab], classes=[0,1]), 
                        average='weighted')
    
    # Por último, la matriz de confusión
    cm[lab] = confusion_matrix(label_val, y_pred[lab])
    
    metrics.append(pd.Series({'precision':precision, 'recall':recall, 
                              'fscore':fscore, 'accuracy':accuracy,
                              'auc':auc}, 
                             name=lab))

metrics = pd.concat(metrics, axis=1)
```

La regresión es una técnica estadística profundamente estudiada que hoy se aplica al machine learning. Existen numerosos casos de éxito de su aplicación en situaciones prácticas. Aunque muchos piensan que solo existen la regresión lineal y la logística, en realidad hay muchos más modelos de regresión, desarrollados para manejar alta dimensionalidad y multicolinealidad en los conjuntos de datos.

En este capítulo se estudió el propósito de algunos de ellos y cómo ayudan a tratar diferentes tipos de datos. Por lo tanto, la selección del algoritmo para tu conjunto de datos depende en gran medida de los resultados del análisis exploratorio de datos (EDA). Scikit-learn y otras librerías proporcionan la implementación de muchos algoritmos de regresión.

Como científico de datos, experimenta con estos algoritmos preimplementados durante el desarrollo del modelo. Selecciona el modelo que tenga mejor desempeño. Si no estás satisfecho con la precisión del modelo, prueba a ajustar finamente sus parámetros y, si aún no funciona, regresa al EDA. Esto ahorra considerablemente tiempo en el desarrollo de modelos.
