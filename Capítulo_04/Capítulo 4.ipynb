## En Pocas Palabras

Este es un algoritmo de aprendizaje supervisado utilizado principalmente para clasificación, pero también puede emplearse para análisis de regresión. Los árboles de decisión imitan el pensamiento humano al tomar decisiones y, por lo tanto, suelen ser fáciles de entender. Podemos representar todo el árbol de manera visual para una mejor comprensión.  
Para motivar el estudio de los árboles de decisión, primero hablaré sobre las áreas y los dominios donde se han utilizado con éxito.

## Amplia gama de aplicaciones 


Estos son varios dominios y casos de uso donde los árboles de decisión se han aplicado con éxito. Se utilizan en el diagnóstico médico de enfermedades, detección de fraudes, retención de clientes, marketing, sector educativo, entre otros. En el ámbito médico, los médicos y profesionales de la salud emplean árboles de decisión para la identificación temprana de condiciones prevenibles, como la diabetes o la demencia. Los bancos utilizan árboles de decisión para identificar comportamientos fraudulentos de clientes anteriores y así prevenir fraudes, ahorrando grandes cantidades en pérdidas. Las empresas utilizan árboles de decisión para la retención de clientes analizando sus compras y lanzando nuevas ofertas basadas en sus hábitos de consumo. También pueden usar árboles de decisión para evaluar los niveles de satisfacción del cliente.

Los negocios observan el desempeño de los competidores en sus productos y servicios y construyen árboles de decisión para la segmentación de clientes, mejorando la precisión de sus nuevas campañas promocionales. Las universidades pueden preseleccionar estudiantes en función de sus méritos, asistencia, calificaciones, etc., mediante la construcción de árboles de decisión. Por lo tanto, existen numerosas aplicaciones de los árboles de decisión. En el pasado, hemos utilizado árboles de decisión para varias de estas aplicaciones porque imitan la capacidad de pensamiento humano al tomar decisiones y también son fácilmente interpretables gracias a su estructura en forma de árbol.  

## Funcionamiento del árbol de decisión  


En el uso del algoritmo de árbol de decisión, intervienen dos procesos:  
• Construcción del propio árbol  
• Recorrido del árbol para inferir un punto de datos no visto  

De los dos anteriores, la construcción del árbol es más complicada, mientras que usar el árbol para la inferencia es relativamente sencillo. Primero hablaré sobre el recorrido y luego mostraré cómo construir un árbol presentando el algoritmo para su construcción.

### Entropía


La entropía es una medida de cuán desordenados están tus datos. Una alta entropía significa que los datos están en caos y una baja entropía significa que los datos están distribuidos uniformemente. Considera nuestro ejemplo anterior, donde tenemos los datos etiquetados: un jugador está “jugando” o “no jugando”. Llamaremos a estos valores “P” para jugando y “NP” para no jugando. La figura 4.3 muestra ambos casos de entropía baja y alta.

La entropía es casi cero cuando el conjunto de datos se aproxima a la homogeneidad y es igual a uno cuando está distribuido equitativamente. Durante la construcción del árbol, tener una baja entropía ayuda a una mejor segregación de las clases y, por lo tanto, crea un mejor árbol para las predicciones.  

Matemáticamente, la entropía se expresa como:  

$$
\text{Entropy}(S) = - P_{\bigcirc\!\!\!\!\times} \log_2 P_{\bigcirc\!\!\!\!\times} \;-\; P_{\bigcirc\!\!\!\! -} \log_2 P_{\bigcirc\!\!\!\! -}
$$

donde $P_{\textcircled{+}}$ es la proporción de ejemplos positivos y $P_{\textcircled{-}}$ es la proporción de ejemplos negativos en $S$.  

Mostraré cómo se aplica esta fórmula cuando explique la construcción del árbol. También definiré otro término llamado **ganancia de información**.

### Ganancia de Información


Durante la construcción del árbol, se utiliza la ganancia de información para generalizar la entropía. Como se mencionó anteriormente, tener baja entropía ayuda a crear modelos de predicción más precisos. Matemáticamente, la ganancia de información se define de la siguiente manera:

$$
\text{Gain}(S, A) = \text{Entropy}(S) - \sum_{v \in \text{Values}(A)} \frac{|S_v|}{|S|} \text{Entropy}(S_v)
$$

donde $\text{Values}(A)$ son todos los posibles valores del atributo $A$ y $S_v$ es el subconjunto de $S$ para el cual el atributo $A$ tiene el valor $v$.

En cada división de un nodo, se calcula la ganancia de información: cuanto mayor sea el valor, mejor será la división, ya que una alta ganancia de información indica baja entropía. Baja entropía significa una distribución más uniforme. La distribución uniforme permite una mejor segregación en clases. Por lo tanto, la ganancia de información puede ser una buena medida para decidir qué atributos son los más relevantes. Un atributo más relevante se usa como nodo y se divide en ramas.

Una característica como el número de tarjeta de crédito puede tener muchos valores distintos, resultando en valores altos de ganancia de información. Esto puede ser una desventaja al trabajar con conjuntos de datos que tienen columnas con muchos valores distintos. Por lo tanto, seguramente se excluirán los números de tarjeta de crédito en el análisis, ya que no se desea generalizar a los clientes en función de sus números de tarjeta.

### Índice de Gini


El índice de Gini se expresa matemáticamente como:  

$$
G = \sum_{i=1}^{C} p(i) \, (1 - p(i))
$$

donde $C$ es el número de clases y $p(i)$ es la probabilidad de seleccionar un punto de datos que pertenezca a la clase $i$.

Podemos simplificar la expresión a:  

$$
G = \sum_{i=1}^{C} p(i) \, * (1 - p(i)) = \sum_{i=1}^{C} p(i) - \sum_{i=1}^{C} p(i)^2 = 1 - \sum_{i=1}^{C} p(i)^2
$$

donde $p(i)$ es la probabilidad de la clase $i$ en el atributo. El primer término de la ecuación anterior, que representa la suma de todas las probabilidades, es igual a 1.

Por lo tanto, el índice de Gini se calcula restando la suma de los cuadrados de las probabilidades de cada clase del número uno:  

$$
\text{Gini}(\text{atributo}) = 1 - \sum_{i=1}^{N} (P_i)^2
$$

$$
\text{Gini}(\text{atributo}) = \sum_{v \in \text{valores}} P_v \, \text{Gini}(\text{atributo}_v)
$$

donde $P_v$ es la probabilidad de los valores en el atributo y $P_i$ es la probabilidad de seleccionar un punto de datos con la clase $i$.

Es sencillo de implementar y favorece la división en particiones más grandes. El índice de Gini varía entre 0 y 1, donde 0 representa pureza en la clasificación y 1 denota distribución aleatoria de los elementos entre varias clases. Un índice de Gini de 0.5 indica que hay una distribución equitativa de elementos entre algunas clases. En otras palabras, esencialmente nos da la probabilidad de clasificar incorrectamente un atributo seleccionado.

Ahora llegaré al punto importante de la construcción del árbol. Tenemos dos implementaciones diferentes para construir el árbol: una basada en la ganancia de información y la otra que utiliza el índice de Gini. Ahora mostraré cómo construir un árbol para estos datos parciales.  
Primero, realizamos los cálculos de entropía para nuestro problema completo, es decir, para la distribución completa de clases o etiquetas en la columna objetivo.

El conjunto de datos tiene 14 instancias positivas (Yes) y 6 negativas (No); por lo tanto, $S = [14^+, 6^-]$:

$$
\text{Entropy}(S) = \text{Entropy}[14^+, 6^-] = - \frac{14}{20} \log_2 \frac{14}{20} - \frac{6}{20} \log_2 \frac{6}{20} \approx 0.88
$$


Esto indica que el conjunto de datos es 88% impuro o no homogéneo. A continuación, medimos la efectividad de cada atributo para clasificar el conjunto de entrenamiento. La medida que utilizaremos se llama ganancia de información o *information gain*.

Para mayor claridad, usemos la fórmula de ganancia de información y calculemos la ganancia del atributo *Wind* a partir del conjunto de datos. El atributo *Wind* tiene los siguientes valores: *Weak* o *Strong*. Por lo tanto, la distribución de clases dentro del atributo *Wind* sería:  

- Wind(Weak) $= W_W = [9^+, 2^-]$, es decir, viento débil tiene 9 instancias de Yes y 2 instancias de No.  
- Wind(Strong) $= W_S = [5^+, 4^-]$, es decir, viento fuerte tiene 5 instancias de Yes y 4 instancias de No.  
- $\text{Entropy}(S) = 0.88$, es decir, la entropía del espacio muestral completo, que calculamos anteriormente.

La ganancia de información para el atributo *Wind* se expresa como:  

$$
\text{Gain}(S, \text{Wind}) = \text{Entropy}(S) - \left[ \frac{11}{20} \text{Entropy}(W) + \frac{9}{20} \text{Entropy}(W_S) \right]
$$


Para viento débil (Weak):  

$$
\text{Entropy}(W_W) = -\frac{9}{11} \log_2 \frac{9}{11} - \frac{2}{11} \log_2 \frac{2}{11} \approx 0.6840
$$

Para viento fuerte (Strong):  

$$
\text{Entropy}(W_S) = -\frac{5}{9} \log_2 \frac{5}{9} - \frac{4}{9} \log_2 \frac{4}{9} \approx 0.9910
$$

Sustituyendo estos valores en la fórmula principal:  

$$
\text{Gain}(S, \text{Wind}) = 0.88 - \left( \frac{11}{20} \cdot 0.6840 + \frac{9}{20} \cdot 0.9910 \right) 
= 0.88 - (0.3762 + 0.44595) \approx 0.0578
$$

Por lo tanto, la ganancia de información del atributo Wind es 0.057.

De manera similar, para el atributo *Outlook* tenemos tres valores: *Sunny* (Soleado), *Rain* (Lluvioso) y *Overcast* (Nublado). La distribución de clases es:  

- Outlook(Sunny) $O_S = [6^+, 2^-]$  
- Outlook(Overcast) $O_O = [4^+, 0^-]$  
- Outlook(Rain) $O_R = [4^+, 4^-]$ 

La ganancia de información para el atributo de *Outlook* se expresará de la siguiente manera. Calculamos las entropías de cada subgrupo:

Para soleado (*Sunny*):

$$
\text{Entropy}(O_S) = -\frac{6}{8} \log_2 \frac{6}{8} - \frac{2}{8} \log_2 \frac{2}{8} \approx 0.8112
$$

Para nublado (*Overcast*):

$$
\text{Entropy}(O_O) = -\frac{4}{4} \log_2 \frac{4}{4} - \frac{0}{4} \log_2 \frac{0}{4} = 0
$$

Para lluvioso (*Rainy*):

$$
\text{Entropy}(O_R) = -\frac{4}{8} \log_2 \frac{4}{8} - \frac{4}{8} \log_2 \frac{4}{8} = 1
$$

Sustituyendo estos valores en la fórmula principal:  

$$
\text{Gain}(S, \text{Outlook}) = 0.88 - \left( \frac{8}{20} \cdot 0.8112 + \frac{4}{20} \cdot 0 + \frac{8}{20} \cdot 1 \right) 
= 0.88 - (0.3245 + 0 + 0.4) \approx 0.1555
$$

De manera similar, calculamos la ganancia de información para el resto de los atributos/características, obteniendo los siguientes resultados:  

- $\text{Gain(Outlook)} = 0.1555$  
- $\text{Gain(Wind)} = 0.0573$  
- $\text{Gain(Temperature)} = 0.0019$  
- $\text{Gain(Humidity)} = 0.0018$ 


El objetivo principal de la ganancia de información es encontrar el atributo/característica más útil para clasificar los datos. El algoritmo ID3 utiliza este atributo como la raíz para construir el árbol de decisión. Luego calcula repetidamente la ganancia para encontrar el siguiente nodo. Actualmente, el atributo más útil es *Outlook*, ya que proporciona más información que los demás atributos. Por lo tanto, *Outlook* será la raíz de nuestro árbol.

Arriba están los puntos de datos con la característica *sunny*. El subconjunto completo tiene ocho entradas con seis instancias de yes y dos instancias de no.  
El espacio muestral de este nodo sería:  

$$
S_{\text{sunny}} = [6^+, 2^-]
$$



La entropía para este subconjunto se daría como:  

$$
\text{Entropy}(S) = \text{Entropy}[6^+, 2^-] = -\frac{6}{8} \log_2 \frac{6}{8} - \frac{2}{8} \log_2 \frac{2}{8} \approx 0.81
$$

Para este subconjunto, tenemos una entropía de clase de 0.81%.  

Empecemos calculando la ganancia de información para el atributo *humidity* (humedad) para el subconjunto tomado del árbol. La distribución para humidity en este caso sería:  

- Humidityhigh $H_H = [6^+, 1^-]$
- Humiditynormal $H_N = [0^+, 1^-]$

Ahora calculamos la ganancia de información para esta característica, usando las siguientes ecuaciones

Ganancia de información para *Humidity* (Humedad):  

$$
\text{Gain}(S, \text{Humidity}) = \text{Entropy}(S) - \left[ \frac{7}{8} \text{Entropy}(H_H) + \frac{1}{8} \text{Entropy}(H_N) \right]
$$

Calculamos las entropías de los subgrupos:  

$$
\text{Entropy}(H_H) = -\frac{6}{7} \log_2 \frac{6}{7} - \frac{1}{7} \log_2 \frac{1}{7} = 0.5915 + 0 \approx 0.5915
$$

$$
\text{Entropy}(H_N) = -\frac{0}{1} \log_2 \frac{0}{1} - \frac{1}{1} \log_2 \frac{1}{1} = 0 + 0 \approx 0
$$

Sustituyendo los valores encontrados en la fórmula principal, obtenemos:  

$$
\text{Gain}(S, \text{Humidity}) = 0.81 - \left( \frac{7}{8} \cdot 0.5915 + \frac{1}{8} \cdot 0 \right) \approx 0.5175
$$

Ahora medimos la ganancia de *Temperature* (Temperatura) y *Wind* (Viento) siguiendo la misma manera que para Gain(S, Humidity):  

- $\text{Gain}(S, \text{Humidity}) = 0.5175$  
- $\text{Gain}(S, \text{Temperature}) = 0.0725$  
- $\text{Gain}(S, \text{Wind}) = 0.0725$

El atributo *Humidity* (Humedad) nos da la mayor información en esta etapa de la iteración.

| D11 | Rain    | Mild       | Normal  | Weak   | Yes  |
| D15 | Rain    | Mild       | High    | Weak   | Yes  |

Arriba están los puntos de datos con la característica *Rain* (Lluvia). El subconjunto completo tiene ocho entradas con cuatro instancias de yes y cuatro instancias de no.  
El espacio muestral de este nodo sería:  
$$
S_{\text{rain}} = S = [4^+,4^-]
$$

La entropía para este subconjunto se daría como:  
$$
\text{Entropy}(S) = \text{Entropy}[4^+,4^-] = - \left( \frac{4}{8} \log_2 \frac{4}{8} + \frac{4}{8} \log_2 \frac{4}{8} \right) = 1
$$

Ya que todos los puntos de datos asociados con *Outlook* (Sunny y Overcast) y *Humidity* están clasificados, las únicas características que quedan son *Wind* (Viento) y *Temperature* (Temperatura) en este caso.  
Calcularemos ganancia de información para estas dos características respectivo al subconjunto del árbol.  

Ahora calculemos ganancia de información para atributo/característica *Wind* (Viento).  
Distribución para *Wind* en el caso del sub-árbol sería:  
- Wind(Weak) $W_W = [3^+,1^-]$  
- Wind(Strong) $W_S = [1^+,3^-]$ 

Ahora calculamos ganancia de información para esta característica, usando la siguiente ecuación:  
$$
\text{Gain}(S, \text{Wind}) = \text{Entropy}(S) - \left[ \frac{4}{8} \text{Entropy}(W_W) + \frac{4}{8} \text{Entropy}(W_S) \right]
$$

$$
\text{Entropy}(W_W) = - \left( \frac{3}{4} \log_2 \frac{3}{4} + \frac{1}{4} \log_2 \frac{1}{4} \right) = 0.3112 + 0.5 = 0.8112
$$

$$
\text{Entropy}(W_S) = - \left( \frac{1}{4} \log_2 \frac{1}{4} + \frac{3}{4} \log_2 \frac{3}{4} \right) = 0.5 + 0.3112 = 0.8112
$$

Sustituimos los valores que encontramos en la fórmula principal, y obtenemos:  
$$
\text{Gain}(S, \text{Wind}) = \text{Entropy}(S) - \left[ \frac{4}{8} \cdot 0.8112 + \frac{4}{8} \cdot 0.8112 \right] = 1 - (0.4056 + 0.4056) = 0.1888
$$

Calculemos la ganancia de información para el atributo/característica *Temperature* (Temperatura).  
La distribución para *Temperature* en el caso del sub-árbol sería:  

- Temperaturehot $T_H = [0^+,0^-]$  
- Temperaturemild $T_M = [3^+,2^-]$  
- Temperaturecool $T_C = [1^+,2^-]$  

Dado que en este subconjunto no hay instancias de *Temperature* hot, tomaremos 0 casos positivos y 0 casos negativos para este cálculo.

Calculamos la ganancia de información para esta característica usando la siguiente ecuación:  

$$
\text{Gain}(S, \text{Temperature}) = \text{Entropy}(S) - \left[ \frac{0}{8} \text{Entropy}(T_H) + \frac{5}{8} \text{Entropy}(T_M) + \frac{3}{8} \text{Entropy}(T_C) \right]
$$

Calculamos las entropías de los subgrupos:  

$$
\text{Entropy}(T_H) = -\frac{0}{0} \log_2 \frac{0}{0} - \frac{0}{0} \log_2 \frac{0}{0} = 0 + 0 = 0
$$

$$
\text{Entropy}(T_M) = -\frac{3}{5} \log_2 \frac{3}{5} - \frac{2}{5} \log_2 \frac{2}{5} = 0.4421 + 0.5287 = 0.9708
$$

$$
\text{Entropy}(T_C) = -\frac{1}{3} \log_2 \frac{1}{3} - \frac{2}{3} \log_2 \frac{2}{3} = 0.5253 + 0.3899 = 0.9182
$$

Sustituimos los valores encontrados en la fórmula principal, y obtenemos:  

$$
\text{Gain}(S, \text{Temperature}) = 1 - \left[ \frac{0}{8} \cdot 0 + \frac{5}{8} \cdot 0.9708 + \frac{3}{8} \cdot 0.9182 \right]
$$

$$
= 1 - (0 + 0.606 + 0.344) = 0.05
$$

Obtenemos la ganancia de información para ambas características:  

- $\text{Gain}(S, \text{Wind}) = 0.1888$  
- $\text{Gain}(S, \text{Temperature}) = 0.05$

El atributo *Wind* (Viento) nos da la mayor información en esta etapa de la iteración para la construcción.  
El nodo después de *Outlook* (Pronóstico) en el descendiente *Rain* (Lluvia) será el nodo *Wind* (Viento).  
En el atributo *Wind*, el descendiente *Weak* (Débil) tiene únicamente ejemplos "Yes", y el descendiente *Strong* (Fuerte) tiene únicamente ejemplos "No". Por lo tanto, ambas características *Weak* y *Strong* pueden converger para clasificar el objetivo (nodo hoja) y no pueden expandirse más.

Ahora que todos los puntos de datos en el conjunto de entrenamiento están clasificados usando el siguiente procedimiento, el árbol final construido sería como se ve en la Fig. 4.6.

### Algoritmo de Construcción de Árbol

Expresamos el proceso completo mediante el siguiente algoritmo:  

- Seleccione la característica que mejor clasifique el conjunto de datos en las clases deseadas y asigne esa característica al nodo raíz.  
- Recorra desde el nodo raíz tomando decisiones relevantes en cada nodo interno, de manera que cada nodo interno clasifique mejor los datos.  
- Vuelva al paso 1 y repita hasta asignar una clase a los datos de entrada.

### Algoritmo de Recorrido de Árbol

Se recorre el árbol para visitar un nodo en un árbol o para imprimir todos los valores que contiene.  
Existen tres maneras de recorrer un árbol. Estas se llaman:  
- In-order (En orden)  
- Pre-order (Preorden)  
- Post-order (Postorden)

En el método de recorrido *in-order*, visitamos primero el subárbol izquierdo, seguido por la raíz y finalmente el subárbol derecho.  
En el recorrido *pre-order*, se visita primero la raíz, seguido por el subárbol izquierdo y finalmente el subárbol derecho.  
En el recorrido *post-order*, primero visitamos el subárbol izquierdo, seguido por el derecho y finalmente la raíz.  

Los tres métodos de recorrido son equivalentes, y se puede elegir cualquiera de ellos sin más consideraciones. Los efectos serían los mismos en los tres casos.

## Implementación

La biblioteca *sklearn* proporciona la implementación de algoritmos de árboles de decisión en dos clases: una para regresión *sklearn.tree.DecisionTreeRegressor* y otra para clasificación *sklearn.tree.DecisionTreeClassifier*.

```python
# class sklearn.tree.DecisionTreeClassifier(*,
# criterion='gini', splitter='best',
# max_depth=None, min_samples_split=2,
# min_samples_leaf=1,
# min_weight_fraction_leaf=0.0,
# max_features=None, random_state=None,
# max_leaf_nodes=None, min_impurity_decrease=0.0,
# class_weight=None, ccp_alpha=0.0)
```

Ambas clases aceptan muchos parámetros; sin embargo, en su mayoría se utilizarán los parámetros predeterminados establecidos por los diseñadores. Usar valores predeterminados para parámetros como *max_depth*, *min_samples_leaf*, etc., que controlan el tamaño del árbol, dará lugar a un árbol completamente desarrollado y sin poda. En conjuntos de datos grandes, esto puede generar árboles potencialmente muy grandes. Entonces, será necesario establecer los valores de estos parámetros para controlar la complejidad y el tamaño de los árboles.

## Proyecto (Regresión)  

Usaré el famoso conjunto de datos de viviendas de Boston proporcionado en el repositorio de aprendizaje automático UCI para estimar el precio de una casa en la región de Boston.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
from sklearn.tree import plot_tree
from sklearn.metrics import r2_score
from sklearn.datasets import fetch_california_housing
import warnings
warnings.filterwarnings('ignore')
```

**NOTA**: `load_boston` ha sido eliminado de *scikit-learn* desde la versión 1.2. En su lugar, usaremos `fetch_california_housing`.

### Cargando el Conjunto de Datos

Como el conjunto de datos está disponible como parte de la biblioteca *sklearn*, se carga en el proyecto simplemente llamando a la función incorporada. Después de cargar el conjunto de datos, cree un *DataFrame* de *Pandas* para su uso posterior.

```python
house_dataset = fetch_california_housing()
house = pd.DataFrame(house_dataset.data, columns=house_dataset.feature_names)
house['MEDV']=house_dataset.target
```

### Preparación de los Conjuntos de Datos

A continuación, se extraen las características (*features*) y el objetivo (*target*).

#### Selección de Características

Los índices de progresión de la enfermedad, indicados por la variable *TRGT*, son nuestra variable objetivo (*target*), y las restantes son las variables de características (*features*) en base a las cuales predeciremos el valor de TRGT.

```python
# Drop Latitude and Longitude from features
X = house.drop(['MEDV'], axis=1)
y = house['MEDV']
```

#### División de Datos

Dividiremos los datos en conjuntos de entrenamiento y prueba en una proporción de 80:20 %. Hemos reservado el 20 % de los datos para pruebas. En este punto, nuestros conjuntos de datos están listos y procederemos a la construcción del modelo.

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

Estructura de los Conjuntos:

```python

print("shape of training set: ",x_train.shape)#shape of training set 
print("shape of test set: ",x_test.shape)#shape of test set
```

### Construcción del Modelo

Para la inicialización del modelo, usaremos *DecisionTreeRegressor* para realizar regresión sobre el conjunto de datos proporcionado. Aplicamos el *DecisionTreeRegressor* sobre el conjunto de datos de entrenamiento con el siguiente código:

```python
model = DecisionTreeRegressor(
    max_depth=5,          # limits tree depth
    max_leaf_nodes=25,    # restricts number of leaf nodes
    min_samples_split=10, # minimum samples required to split an internal node
    min_samples_leaf=5,   # minimum samples required to be at a leaf node
    random_state=123
)
```

La función *DecisionTreeRegressor* acepta varios parámetros; aquí solo hemos usado sus valores predeterminados.  
Entrenamos el modelo llamando a su método *fit*.

```python
model.fit(x_train,y_train)
```

La biblioteca *sklearn* también proporciona una función llamada *DecisionTreeClassifier*, que se usa si la columna objetivo (*target*) tiene valores discretos.  
En nuestro caso, el objetivo tiene valores continuos.

### Evaluación del Rendimiento (Predicción y Precisión)

Después de entrenar el modelo, se realiza la inferencia sobre el conjunto de prueba llamando al método *predict* y se evalúa la precisión:

```python
y_pred=model.predict(x_test)
```

```python
print('Accuracy on test set: {:.2f} %'.format(model.score(x_test,y_test)*100))
```

En regresión, R-Cuadrado (*R-Squared*) es la medida más popular de la bondad de ajuste. Un valor de *R-Squared* más cercano a 1 indica un mejor ajuste. En mi ejecución de prueba me dio alrededor del 87 % de precisión.

```python
print("R-Squared = {}".format(r2_score(y_test, y_pred)))
```

También puede verificar el *R2-score*, que en mi caso fue 86.79 %.  
En regresión, *R-Squared* es la medida más popular de la bondad de ajuste. Un valor de *R-Squared* más cercano a 1 indica un mejor ajuste.

## Proyecto (Clasificador)  

Para este proyecto, he utilizado el conjunto de datos de cáncer de mama proporcionado en la biblioteca sklearn.  
Es un conjunto de datos pequeño (569 instancias) y, por lo tanto, ideal para la visualización de árboles de decisión. El conjunto contiene 30 características que determinan el tipo de cáncer (maligno o benigno), que será nuestra variable objetivo. Por lo tanto, se trata de un problema de clasificación binaria.  
Se mostrarán tanto los enfoques de *gini* como de *entropy* para construir el clasificador.  

El conjunto de datos se carga usando el siguiente código:

```python
from sklearn import datasets

df = datasets.load_breast_cancer()

df.data.shape
```

```python
print("Features: ", df.feature_names)
```

```python
print("Labels: ", df.target_names)
```

### Creación de conjuntos de datos de entrenamiento o prueba

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.data, df.target, test_size=0.3,random_state=109) 
print("shape of X_train=",X_train.shape)
print("shape of y_train=",y_train.shape)
print("shape of X_test=",X_test.shape)
print("shape of y_test=",y_test.shape)
```

### Clasificador de árbol de decisión (Gini)

#### Construcción del Modelo

Puedes verificar los nombres de las características examinando el valor del atributo *feature_names* y las etiquetas examinando el valor de *target_names*. #### Predicción y Evaluación de Modelo

```python
y_pred_gini = classifier.predict(X_test)
```

A continuación, realiza la predicción sobre el conjunto de prueba y imprime la puntuación de precisión (*accuracy score*) como se hizo anteriormente. Este fue el resultado obtenido en mi ejecución de prueba:
